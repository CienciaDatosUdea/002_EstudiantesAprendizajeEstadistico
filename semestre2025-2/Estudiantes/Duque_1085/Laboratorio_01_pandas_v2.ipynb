````python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

import re
````

2. Renombrar las columnas en formato PascalCase.

````python
data = pd.read_csv('https://archive.ics.uci.edu/static/public/17/data.csv')

def Pascal_Case(Name):
  Name_list = Name.split("_")
  New_Name=""

  for i in Name_list:
    New_Name += i.capitalize()

  return New_Name

data.columns = [Pascal_Case(c) for c in data.columns]
````

8. Normalizar cada columna respecto a su media y desviación estándar utilizando la fórmula:  
   `(x - mean(x)) / std(x)`

````python
data["Area1"].mean()

for col in [i for i in data.columns if i != "Diagnosis"]:
    data[col] = (data[col] - data[col].mean()) / data[col].std()

data
````


9. Agrupar características similares y calcular su promedio.
   Para las siguientes características:

```python
['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean', 'SmoothnessMean', 'CompactnessMean', 'ConcavityMean', 'ConcavePointsMean', "SymmetryMean", "FractalDimensionMean"]
```

Usa expresiones regulares para identificar patrones comunes, como `radius1`, `radius2`,`radius3`, etc., y calcular los promedios. Por ejemplo:

```python
re.match(r'^[a-zA-Z_]+', "holamundo12341").group(0)
```

También puedes utilizar el método `startswith()` para buscar estos patrones en los nombres de las columnas.

````python
#creemos una lista para contener los grupos

groups = []

for col in data.columns[1:-2]:
    groups.append(re.match(r'^[a-zA-Z_]+', col).group(0))

groups = list(set(groups))

mean_data = data[["Id", "Diagnosis", "DiagnosisNumeric"]].copy()

for group in groups:
    mean_data.loc[:, f"{group}Mean"] = data.filter(like=group).mean(axis=1)

mean_data
````

Usa el rango intercuartílico (IQR) para definir los outliers. Los valores fuera del rango [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] se consideran outliers. Ejemplo de cómo eliminar estos valores:

````python
plt.figure(figsize=(10,5))
sns.boxplot(data = mean_data, y = "RadiusMean", hue="Diagnosis" )
plt.title("Boxplot sobre los datos medios")
plt.show()
````

````python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR)))].copy()

A = mean_data[mean_data["Diagnosis"] == "M"].copy().reset_index(drop=True)
B = mean_data[mean_data["Diagnosis"] == "B"].copy().reset_index(drop=True)

A = remove_outliers(A, "RadiusMean")
B = remove_outliers(B, "RadiusMean")

mean_data_filt = pd.concat([A, B], axis=0).reset_index(drop=True)

plt.figure(figsize=(10,5))
sns.boxplot(data = mean_data_filt, y = "RadiusMean", hue="Diagnosis" )
plt.title("Boxplot sobre los datos medios, sin outliners")
plt.show()

mean_data_filt
````
