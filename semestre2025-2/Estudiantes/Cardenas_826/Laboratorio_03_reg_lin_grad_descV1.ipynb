# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y  | X_1  |
|--------|---|----------|
|$1$     |$Y^{1}$ | $X_1^{1}$|
|$2$     |$Y^{2}$ | $X_1^{2}$|
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$m$     |$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

## SOLUCIÓN

1. Dataframe

````python
import pandas as pd
import numpy as np
import sympy as sp

m = sp.symbols('m')

data = {
    'Entrenamiento': [0, 1, 2, 3, 4, m],
    'Y': [0, 1, 2, 3, 4, 5],
    'X_1': [0, 1, 2, 3, 4, 5]
}

# Dataframe
df = pd.DataFrame(data)
df
````

2. Función de Coste

Se define la función de Coste considerando que

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

para la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$

````python
# --- Función de coste ---

def funcion_coste(X, Y, theta_0, theta_1):
    m = len(Y)  # N ejemplos de entrenamiento

    h = theta_0 + (theta_1 * X)

    # Función de coste
    J = (1 / (2 * m)) * np.sum((h - Y)**2)

    return J
````

3. Gráfico de función de Coste considerando un $\theta_0 = 0$ y variando los valores de $\theta_1$

````python
import matplotlib.pyplot as plt

# Datos
X_data = df['X_1'].values
Y_data = df['Y'].values

θ_0_fijo = 0.0
θ_1_val = np.linspace(-2, 4, 100)

# Cálculo
costes = np.array([funcion_coste(X_data, Y_data, θ_0_fijo, t1) for t1 in θ_1_val])

# Valor mínimo
min_coste = np.min(costes)
idx_min_coste = np.argmin(costes)   # Indice
optimal_θ_1 = θ_1_val[idx_min_coste]

# 3. Gráfico Función de coste J(θ_0=0, θ_1)
plt.figure(figsize=(8, 5))
plt.title(r'Función de Coste $J(θ_0=$'f'{θ_0_fijo}'r', $θ_1$) vs $θ_1$')
plt.plot(θ_1_val, costes, color='blue', linewidth=2)
plt.scatter(optimal_θ_1, min_coste, color='red', marker='o', s=100,
            label=f'Mínimo: J={min_coste:.4f} en θ_1={optimal_θ_1:.2f}')
plt.xlabel(r'$θ_1$')
plt.ylabel(r'$J(θ_0, θ_1)$')
plt.legend()
plt.grid(True)
plt.show()
````

Ecuación de Regresión

````python
# Ecuación de regresión obtenida
x_line = np.array([min(X_data), max(X_data)])
y_line = θ_0_fijo + optimal_θ_1 * x_line

# Gráfico Ecuación de Regresión
plt.figure(figsize=(10, 6))
plt.title('Regresión Lineal sobre los Datos Originales')
plt.plot(x_line, y_line, color='red', linewidth=2,
            label=f'Regresión: h(X) = {θ_0_fijo:.2f} + {optimal_θ_1:.2f}X')
plt.scatter(X_data, Y_data, color='green', marker='x', s=100,
            label=r'Datos de Entrenamiento ($X_1$, Y)')
plt.xlabel(r'$X_1$')
plt.ylabel(r'$Y$')
plt.grid(True)
plt.legend()
plt.show()
````

4. Gráfico 3D Función de Coste

````python
# Valores para θ0 y θ1
θ_0_vals = np.linspace(-2, 4, 100) # De -3 a 3 para θ_0
θ_1_vals = np.linspace(-4, 3, 100) # De -3 a 3 para theta_1

# Malla
t0, t1 = np.meshgrid(θ_0_vals, θ_1_vals)

J = np.zeros(t0.shape)

for i in range(t0.shape[0]):
    for j in range(t0.shape[1]):
        J[i, j] = funcion_coste(X_data, Y_data, t0[i, j], t1[i, j])

# Gráfico
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(t0, t1, J, cmap='viridis', alpha=0.7)

# Punto mínimo óptimo (debería ser (0, 1, 0))
ax1.scatter(0, 1, 0, color='red', marker='o', s=50, label='Mínimo Óptimo')

ax1.set_xlabel(r'$θ_0$ (Intercepto)')
ax1.set_ylabel(r'$θ_1$ (Pendiente)')
ax1.set_zlabel(r'$J(θ_0, θ_1)$')
ax1.set_title(r'Superficie de la Función de Coste $J$')
ax1.legend()

# Subgráfico 2: Contorno 2D
ax2 = fig.add_subplot(122)

# Gráfico de contorno (200 niveles)
ax2.contour(t0, t1, J, 200, linestyles='solid', cmap='viridis')

# Punto mínimo óptimo (0, 1)
ax2.plot(0, 1, 'ro', label='Mínimo Óptimo')

ax2.set_xlabel(r'$θ_0$ (Intercepto)')
ax2.set_ylabel(r'$θ_1$ (Pendiente)')
ax2.set_title(r'Gráfico de Contorno de $J$')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
````

5. Encontrar la expresión teórica para la función de coste en el caso 1D y 2D.
---

La función de coste es:

$$J(\theta_0, \theta_1)=\frac{1}{2m} \sum_{i=1}^m ( \theta_0 + \theta_1 X^{(i)}- y^{(i)})^2$$


1.  **En el Caso 1D: Asumiendo $\theta_0 = 0$**

Cuando se asume que $\theta_0 = 0$ (es decir, la recta de regresión pasa por el origen), la hipótesis es $h(X) = \theta_1 X$. La función de coste se simplifica a:

$$J(\theta_1)=\frac{1}{2m} \sum_{i=1}^m ( \theta_1 X^{(i)}- y^{(i)})^2$$

Para encontrar el $\theta_1$ óptimo ($\theta_1^*$), derivamos $J(\theta_1)$ con respecto a $\theta_1$ e igualamos a cero:

- **Derivada Parcial:**

$$\frac{\partial J(\theta_1)}{\partial \theta_1} = \frac{1}{2m} \sum_{i=1}^m 2 (\theta_1 X^{(i)} - y^{(i)}) X^{(i)} = 0$$

- **Despejando $\theta_1$:** Eliminando las constantes $\frac{1}{m}$ y 2, nos queda:

$$\sum_{i=1}^m (\theta_1 X^{(i)} - y^{(i)}) X^{(i)} = 0$$

$$\sum_{i=1}^m (\theta_1 (X^{(i)})^2 - y^{(i)} X^{(i)}) = 0$$

$$\theta_1 \sum_{i=1}^m (X^{(i)})^2 = \sum_{i=1}^m y^{(i)} X^{(i)}$$

- Finalmente, la **expresión Teórica (1D)** nos queda:

$$\theta_1^* = \frac{\sum_{i=1}^m X^{(i)} y^{(i)}}{\sum_{i=1}^m (X^{(i)})^2}$$


2. **Caso 2D: Sin Restricción (Regresión Lineal Simple)**

Para encontrar los valores óptimos de $\theta_0$ y $\theta_1$ ($\theta_0^*$ y $\theta_1^*$), derivamos $J(\theta_0, \theta_1)$ con respecto a $\theta_0$ y $\theta_1$ por separado e igualamos a cero, lo que genera un sistema de dos ecuaciones.

- **Derivada respecto a $\theta_0$:**

$$\frac{\partial J}{\partial \theta_0} = \frac{1}{m} \sum_{i=1}^m ( \theta_0 + \theta_1 X^{(i)}- y^{(i)}) = 0$$

$$\sum_{i=1}^m y^{(i)} = \sum_{i=1}^m (\theta_0 + \theta_1 X^{(i)})$$

$$m \theta_0 + \theta_1 \sum_{i=1}^m X^{(i)} = \sum_{i=1}^m y^{(i)}$$

- Dividiendo por $m$:

$$\theta_0 + \theta_1 \bar{X} = \bar{Y}$$

Esto nos da la Expresión Teórica para $\theta_0$:

$$\theta_0^* = \bar{Y} - \theta_1^* \bar{X}$$Donde $\bar{Y}$ y $\bar{X}$ son las medias de $Y$ y $X$, respectivamente.

- **Derivada respecto a $\theta_1$:**

$$\frac{\partial J}{\partial \theta_1} = \frac{1}{m} \sum_{i=1}^m ( \theta_0 + \theta_1 X^{(i)}- y^{(i)}) X^{(i)} = 0$$

Sustituyendo $\theta_0$ por la expresión de arriba ($\theta_0 = \bar{Y} - \theta_1 \bar{X}$):

$$\sum_{i=1}^m ( (\bar{Y} - \theta_1 \bar{X}) + \theta_1 X^{(i)}- y^{(i)}) X^{(i)} = 0$$

$$\sum_{i=1}^m ( (\theta_1 X^{(i)} - \theta_1 \bar{X}) - (y^{(i)} - \bar{Y}) ) X^{(i)} = 0$$

Factorizando $\theta_1$:

$$\sum_{i=1}^m \theta_1 (X^{(i)} - \bar{X}) X^{(i)} = \sum_{i=1}^m (y^{(i)} - \bar{Y}) X^{(i)}$$

Esto conduce a la Expresión Teórica para $\theta_1$:$$\theta_1^* = \frac{\sum_{i=1}^m (X^{(i)} - \bar{X}) (y^{(i)} - \bar{Y})}{\sum_{i=1}^m (X^{(i)} - \bar{X})^2}$$

Esta es la conocida Ecuación Normal para la regresión lineal simple, que minimiza el Error Cuadrático Medio.

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

**Algoritmo de Descenso del Gradiente para $f(x) = (x-4)^2$**

La función objetivo es $f(x) = (x-4)^2$ en donde el mínimo teórico de esta función es en $x=4$, haciendo que $f(x)=0$.

1. Necesitamos la derivada (el gradiente) de la función, $\frac{\mathrm{d}f(x)}{\mathrm{d}x}$, para aplicarla en la regla de actualización:

$$f(x) = (x-4)^2$$

$$\frac{\mathrm{d}f(x)}{\mathrm{d}x} = 2(x-4)$$

2. La regla de actualización del algoritmo es:

$$x_{i+1} = x_{i} - \alpha \frac{\mathrm{d}f(x_i)}{\mathrm{d}x}$$

Sustituyendo la derivada:

$$x_{i+1} = x_{i} - \alpha (2(x_i - 4))$$

````python
def gradiente_descendente(funcion, derivada, x_inicial, alpha, epsilon=1e-4, max_iter=1000):
    """
    Implementa el algoritmo del Descenso del Gradiente para encontrar el mínimo.

    Parámetros:
    funcion: Función objetivo f(x).
    derivada: Derivada de la función objetivo df(x)/dx.
    x_inicial (float): Valor inicial (omega_i).
    alpha (float): Tasa de aprendizaje (learning rate).
    epsilon (float): Criterio de parada (error de diferencia entre iteraciones).
    max_iter (int): Número máximo de iteraciones.

    Retorna:
    tuple: (x_minimo, coste_minimo, historial_x, historial_coste, iteraciones)
    """
    x = x_inicial
    historial_x = [x]
    historial_coste = [funcion(x)]

    for i in range(max_iter):
        # 1. Calcular el gradiente (derivada)
        gradiente = derivada(x)

        # 2. Aplicar la regla de actualización
        x_nuevo = x - (alpha*gradiente)

        # 3. Criterio de parada: si el cambio en x es menor a epsilon
        if np.abs(x_nuevo - x) < epsilon:
            print(f"Convergencia alcanzada en {i+1} iteraciones.")
            break

        # 4. Actualizar el valor y guardar historial
        x = x_nuevo
        historial_x.append(x)
        historial_coste.append(funcion(x))

    return x, funcion(x), historial_x, historial_coste, i + 1

# Funciones de prueba
def f_x(x):
    """Función objetivo."""
    return (x - 4)**2

def df_dx(x):
    """Derivada de la función (el gradiente)."""
    return 2 * (x - 4)
````

````python
# --- Pruebas del Algoritmo con diferentes alphas ---

x_inicio = 0.0 # Valor inicial (omega_i)
epsilon_target = 1e-4

alphas = [0.1, 0.5, 0.99] # Tres valores de alpha a probar

resultados = {}

for alpha in alphas:
    print(f"\n--- Probando con alpha = {alpha} ---")
    x_min, coste_min, hist_x, hist_J, iters = gradiente_descendente(
        f_x, df_dx, x_inicio, alpha, epsilon=epsilon_target)
    resultados[alpha] = {
        'x_min': x_min,
        'coste_min': coste_min,
        'iters': iters,
        'hist_x': hist_x,
        'hist_J': hist_J}
    print(f"Mínimo encontrado en x = {x_min:.4f}")
    print(f"Valor mínimo de f(x) = {coste_min:.4e}")
    print(f"Total de iteraciones: {iters}")
````

````python
plt.figure(figsize=(12, 5))

# Gráfico 1: Valor de x a lo largo de las iteraciones
plt.subplot(1, 2, 1)
for alpha, res in resultados.items():
    plt.plot(res['hist_x'], label=f'$\\alpha={alpha}$')
plt.axhline(y=4, color='r', linestyle='--', label='Mínimo Teórico (x=4)')
plt.title('Convergencia de $x$ hacia el mínimo')
plt.xlabel('Iteraciones')
plt.ylabel('$x$')
plt.legend()
plt.grid(True)

# Gráfico 2: Coste f(x) a lo largo de las iteraciones
plt.subplot(1, 2, 2)
for alpha, res in resultados.items():
    plt.plot(res['hist_J'], label=f'$\\alpha={alpha}$')
plt.title('Descenso del Coste $f(x)$')
plt.xlabel('Iteraciones')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

Para aplicar el Descenso del Gradiente a una función multivariable y no lineal se requiere calcular las derivadas parciales (el gradiente) para ambos parámetros ($x$ y $y$).

La función es:

$$F(x,y) = \sin \left( \frac{1}{2}x^2-\frac{1}{4}y^2 +3\right) \cos (2x+1-e^y)$$

El gradiente $\nabla F(x,y)$ es el vector de las derivadas parciales. Se utilizará la regla de la cadena.

1. **Derivada Parcial con respecto a $x$ ($\frac{\partial F}{\partial x}$):**

$$\frac{\partial F}{\partial x} = \cos \left( \frac{1}{2}x^2-\frac{1}{4}y^2 +3\right) \cdot (x) \cdot \cos (2x+1-e^y) + \sin \left( \frac{1}{2}x^2-\frac{1}{4}y^2 +3\right) \cdot (-\sin (2x+1-e^y)) \cdot (2)$$

Simplificando:

$$\frac{\partial F}{\partial x} = x \cos(A) \cos(B) - 2 \sin(A) \sin(B)$$

Donde: $A = \frac{1}{2}x^2-\frac{1}{4}y^2 +3$ y $B = 2x+1-e^y$.


2. **Derivada Parcial con respecto a $y$ ($\frac{\partial F}{\partial y}$):**

$$\frac{\partial F}{\partial y} = \cos \left( \frac{1}{2}x^2-\frac{1}{4}y^2 +3\right) \cdot (-\frac{1}{2}y) \cdot \cos (2x+1-e^y) + \sin \left( \frac{1}{2}x^2-\frac{1}{4}y^2 +3\right) \cdot (-\sin (2x+1-e^y)) \cdot (-e^y)$$

$$$$Simplificando:$$\frac{\partial F}{\partial y} = -\frac{1}{2}y \cos(A) \cos(B) + e^y \sin(A) \sin(B)$$Donde: $A = \frac{1}{2}x^2-\frac{1}{4}y^2 +3$ y $B = 2x+1-e^y$.


Para graficar, se selecciona un punto de inicio ($x_0, y_0$) en el valle, y se aplican las fórmulas de actualización:

$$x_{i+1} = x_{i} - \alpha \frac{\partial F}{\partial x} \quad \text{y} \quad y_{i+1} = y_{i} - \alpha \frac{\partial F}{\partial y}$$

````python
# Funciones
def F(x, y):
    """Función F(x, y) a minimizar."""
    A = (0.5*x**2) - (0.25*y**2) + 3
    B = 2*x + 1 - np.exp(y)
    return np.sin(A) * np.cos(B)

def gradiente_F(x, y):
    """Calcula el vector gradiente [dF/dx, dF/dy]."""
    A = (0.5*x**2) - (0.25*y**2) + 3
    B = 2*x + 1 - np.exp(y)

    sin_A, cos_A = np.sin(A), np.cos(A)
    sin_B, cos_B = np.sin(B), np.cos(B)
    exp_y = np.exp(y)

    # dF/dx (Derivada parcial respecto a x)
    dF_dx = (x*cos_A*cos_B) - (2*sin_A*sin_B)

    # dF/dy (Derivada parcial respecto a y)
    dF_dy = (-0.5*y*cos_A*cos_B) + (exp_y*sin_A*sin_B)

    return np.array([dF_dx, dF_dy])
````

**Descenso del Gradiente para el Mínimo**

Al observar el mapa de contorno, se ve que hay múltiples mínimos locales (valles de color azul oscuro). Se selecciona un punto de inicio y una tasa de aprendizaje ($\alpha$).

````python
# --- 3. Implementación del Descenso del Gradiente ---

def gradiente_descendente_mult(F, gradiente_F, x_start, y_start, alpha=0.01,
                               max_iter=1000, epsilon=1e-6):

    params = np.array([x_start, y_start])
    historial = [params.copy()]

    for i in range(max_iter):
        # Se calcula el gradiente en el punto actual
        grad = gradiente_F(params[0], params[1])

        # Se aplica la regla de actualización
        params_nuevo = params - alpha * grad

        # Criterio de parada: si el cambio en el parámetro es menor a epsilon
        if np.linalg.norm(params_nuevo - params) < epsilon:
            print(f"Convergencia alcanzada después de {i+1} iteraciones.")
            break

        params = params_nuevo
        historial.append(params.copy())

    return params, F(params[0], params[1]), np.array(historial), i + 1

# Parámetros (Se selecciona cerca de un valle visible en el contorno)
x_inicio = -2
y_inicio = 0.5
alpha = 0.005   # Tasa de aprendizaje ajustada
epsilon_target = 1e-6

# Se ejecuta
min_params, min_coste, historial, iters = gradiente_descendente_mult(
    F, gradiente_F, x_inicio, y_inicio, alpha=alpha, epsilon=epsilon_target)

print("\n--- Resultados del Descenso del Gradiente ---")
print(f"Punto Inicial (x, y): ({x_inicio:.4f}, {y_inicio:.4f})")
print(f"Tasa de Aprendizaje (alpha): {alpha}")
print(f"Iteraciones: {iters}")
print(f"Mínimo Local Encontrado en (x, y): ({min_params[0]:.4f}, {min_params[1]:.4f})")
print(f"Valor Mínimo de F(x, y): {min_coste:.6f}")
````

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

8. **Regresión Lineal con Descenso del Gradiente**

El objetivo es encontrar los parámetros óptimos $(\theta_0, \theta_1)$ que minimicen la función de coste $J(\theta_0, \theta_1)$ para los datos dados.La hipótesis es: $h(X) = \theta_0 + \theta_1 X$.

````python
# Generación de los datos (con ruido)
np.random.seed(42) # Para reproducibilidad
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2*X + 0.02*np.random.random(100)

def gradiente_coste(X, Y, theta_0, theta_1):
    """Calcula el vector gradiente [dJ/d(theta_0), dJ/d(theta_1)]"""
    m = len(Y)
    h_x = theta_0 + theta_1 * X
    error = h_x - Y

    # Derivadas
    dJ_dtheta0 = (1/m) * np.sum(error)
    dJ_dtheta1 = (1/m) * np.sum(error * X)

    return np.array([dJ_dtheta0, dJ_dtheta1])

def descenso_gradiente_regresion(X, Y, alpha=0.01, max_iter=1000, epsilon=1e-6):
    """Algoritmo de Descenso del Gradiente para Regresión Lineal."""
    m = len(Y)

    # Inicialización aleatoria de parámetros
    theta0 = np.random.randn()
    theta1 = np.random.randn()

    historial_coste = []

    for i in range(max_iter):
        grad = gradiente_coste(X, Y, theta0, theta1)

        theta0_ant = theta0
        theta1_ant = theta1

        # Regla de actualización
        theta0 = theta0 - alpha * grad[0]
        theta1 = theta1 - alpha * grad[1]

        coste_actual = np.sum((theta0 + theta1 * X - Y)**2) / (2 * m)
        historial_coste.append(coste_actual)

        # Criterio de parada
        if np.abs(theta0 - theta0_ant) < epsilon and np.abs(theta1 - theta1_ant) < epsilon:
            print(f"Convergencia alcanzada en {i+1} iteraciones.")
            break

    return theta0, theta1, historial_coste, i + 1

alpha_gd = 0.5 # Tasa de aprendizaje
theta0_gd, theta1_gd, coste_hist, iters_gd = descenso_gradiente_regresion(X, y,
                                    alpha=alpha_gd, max_iter=2000, epsilon=1e-6)

print("\n--- Resultados del Descenso del Gradiente (GD) ---")
print(f"Iteraciones: {iters_gd}")
print(f"Intercepto (theta_0): {theta0_gd:.4f}")
print(f"Pendiente (theta_1): {theta1_gd:.4f}")
````

````python
# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos con Ruido')

# Línea de regresión obtenida por Descenso del Gradiente
y_pred_gd = theta0_gd + theta1_gd * X
plt.plot(X, y_pred_gd, color='red', linewidth=2,
         label=f'Regresión GD: y = {theta0_gd:.4f} + {theta1_gd:.4f}X')

# Línea de regresión Teórica (0.2 + 0.2X)
plt.plot(X, 0.2 + 0.2*X, color='green', linestyle='--', linewidth=1,
         label='Regresión Teórica (sin ruido)')

plt.title('Regresión Lineal Simple con Descenso del Gradiente')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
````
