# Regresi贸n lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$

|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |

|$m$     |$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ par谩metros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan m谩s "cercano" a $Y$ para cada $X$.

Para la optimizaci贸n, vamos a definir la funci贸n de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, as铆:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la funci贸n de coste, que permite obtener los valores m谩s cercanos,  esta minimizaci贸n podr谩 ser realizada a trav茅s de diferentes m茅todos el m谩s conocido es el gradiente descendente.

# Gradiente Descendente


Para determinar el m铆nimo,  a trav茅s del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un n煤mero aleatorios inicial $\omega_i$
- Para descender al m铆nimo de la funci贸n, encontremos un  valor para el cual
el deriva de la funci贸n descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el m铆nimo de la funci贸n

# Modelo de *machine learning*: Soluci贸n general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

## **1. Supongamos que un conjunto de caracter铆sticas y datos etiquetados estan dados de la siguiente manera:**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# DataFrame
data = {
    "Entrenamiento": [0, 1, 2, 3, 4, 5],
    "Y": [0, 1, 2, 3, 4, 5],
    "X_1": [0, 1, 2, 3, 4, 5]
}

df = pd.DataFrame(data)

df
````

## **2. Crear una funci贸n para calcular la funci贸n de coste.**

````python
def funcion_coste(theta_0, theta_1, X, y):
    m = len(y)
    y_pred = theta_0 + theta_1 * X
    J = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
    return J
````

## **3. Asumiendo que $_0=0$, represente gr谩ficamente la funci贸n de costo para diferentes valores de $_1$. Determine el valor m铆nimo de la funci贸n de costo y, con este valor, grafique la ecuaci贸n de regresi贸n obtenida sobre los datos del dataframe que ha construido.**

````python
# Datos
X = df["X_1"].values
y = df["Y"].values

# Rango de valores para theta_1
theta_1_vals = np.linspace(-1, 3, 200)
costos = []

# Calcular funci贸n de costo para cada theta_1 (theta_0 = 0)
for t1 in theta_1_vals:
    costos.append(funcion_coste(0, t1, X, y))

costos = np.array(costos)

# Encontrar el m铆nimo
theta_1_min = theta_1_vals[np.argmin(costos)]

# Gr谩fica de la funci贸n de costo
plt.figure()
plt.plot(theta_1_vals, costos)
plt.xlabel(r"$\theta_1$")
plt.ylabel(r"$J(\theta_1)$")
plt.title("Funci贸n de costo para distintos valores de $\\theta_1$")
plt.grid()
plt.show()

# Gr谩fica de la regresi贸n
plt.figure()
plt.scatter(X, y)
plt.plot(X, theta_1_min * X, label=f"$y = {theta_1_min:.2f}x$")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Recta de regresi贸n")
plt.legend()
plt.grid()
plt.show()
````

## **4. Suponiendo ahora que los datos de entrenamiento vienen dados de la siguiente manera: Encontrar la funci贸n de coste para diferentes valores de $\theta_0$,$\theta_1$.**


````python
# Datos
X = df["X_1"].values
y = df["Y"].values

# Rango de valores para theta_0 y theta_1
theta_0_vals = np.linspace(-1, 1, 100)
theta_1_vals = np.linspace(-1, 3, 100)

t0, t1 = np.meshgrid(theta_0_vals, theta_1_vals)
J = np.zeros(t0.shape)

# Evaluar la funci贸n de coste en cada punto del grid
for i in range(t0.shape[0]):
    for j in range(t0.shape[1]):
        J[i, j] = funcion_coste(t0[i, j], t1[i, j], X, y)

# Gr谩fico 3D de la funci贸n de coste
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(t0, t1, J, alpha=0.8)
ax.contour(t0, t1, J, 200, linestyles="solid")
ax.set_xlabel(r"$\theta_0$")
ax.set_ylabel(r"$\theta_1$")
ax.set_zlabel(r"$J(\theta_0,\theta_1)$")
ax.set_title("Superficie de la funci贸n de coste")
plt.show()

# Gr谩fico de contorno
plt.figure()
plt.contour(t0, t1, J, 50)
plt.xlabel(r"$\theta_0$")
plt.ylabel(r"$\theta_1$")
plt.title("Curvas de nivel de la funci贸n de coste")
plt.show()
````

6. Construya un algoritmo en el que emplee el gradiente descendente para determinar el m铆nimo de una funci贸n, determine el m铆nimo con una error epsilon de $1E^{-4}$, pruebe su algoritmo para $f(x)=(x-4)^{2}$ y al menos 3 valores de $\alpha$.

````python
# Definici贸n de la funci贸n y su derivada
def f(x):
    return (x - 4)**2

def df(x):
    return 2 * (x - 4)

# Algoritmo de gradiente descendente
def gradiente_descendente(x0, alpha, epsilon=1e-4, max_iter=10000):
    x = x0
    for _ in range(max_iter):
        x_new = x - alpha * df(x)
        if abs(x_new - x) < epsilon:
            break
        x = x_new
    return x, f(x)

# Valores iniciales y tasas de aprendizaje
x0 = 0
alphas = [0.1, 0.2, 0.5]

# Ejecutar el algoritmo para distintos valores de alpha
resultados = {}
for alpha in alphas:
    minimo, valor_min = gradiente_descendente(x0, alpha)
    resultados[alpha] = (minimo, valor_min)

resultados
````

## **7. Encontrar el m铆nimo de la siguiente funci贸n a trav茅s del m茅todo del gradiente descendente**

````python
# Definici贸n de la funci贸n
def F(x, y):
    return np.sin(0.5*x**2 - 0.25*y**2 + 3) * np.cos(2*x + 1 - np.exp(y))

# Gradientes num茅ricos
def grad_F(x, y, h=1e-5):
    dFx = (F(x + h, y) - F(x - h, y)) / (2*h)
    dFy = (F(x, y + h) - F(x, y - h)) / (2*h)
    return np.array([dFx, dFy])

# Gradiente descendente
def gradiente_descendente_2D(x0, y0, alpha=0.01, epsilon=1e-4, max_iter=10000):
    x, y = x0, y0
    for _ in range(max_iter):
        grad = grad_F(x, y)
        x_new = x - alpha * grad[0]
        y_new = y - alpha * grad[1]
        if np.linalg.norm([x_new - x, y_new - y]) < epsilon:
            break
        x, y = x_new, y_new
    return x, y, F(x, y)

# Grilla para graficar
x_vals = np.linspace(-3, 3, 200)
y_vals = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x_vals, y_vals)
Z = F(X, Y)

# Gr谩fica 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, alpha=0.8)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("F(x,y)")
ax.set_title("Superficie 3D de la funci贸n")
plt.show()

# Mapa de contorno
plt.figure()
plt.contour(X, Y, Z, 50, cmap='viridis')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Mapa de contorno de F(x,y)")
plt.show()

# Gradiente descendente
x_min, y_min, f_min = gradiente_descendente_2D(0.5, 0.5)

x_min, y_min, f_min
````

## **8. Empleando los siguientes datos y las herramientas desarrolladas en los apartados anteriores, construya un algoritmo que permita determinar una regresi贸n lineal.**

````python
# Datos
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)

# Funci贸n de costo
def funcion_coste(theta_0, theta_1, X, y):
    m = len(y)
    y_pred = theta_0 + theta_1 * X
    return (1 / (2 * m)) * np.sum((y_pred - y) ** 2)

# Gradientes de la funci贸n de costo
def gradientes(theta_0, theta_1, X, y):
    m = len(y)
    y_pred = theta_0 + theta_1 * X
    dtheta_0 = (1 / m) * np.sum(y_pred - y)
    dtheta_1 = (1 / m) * np.sum((y_pred - y) * X)
    return dtheta_0, dtheta_1

# Gradiente descendente
theta_0, theta_1 = 0.0, 0.0
alpha = 0.1
epsilon = 1e-6
max_iter = 10000

for _ in range(max_iter):
    d0, d1 = gradientes(theta_0, theta_1, X, y)
    theta_0_new = theta_0 - alpha * d0
    theta_1_new = theta_1 - alpha * d1

    if abs(theta_0_new - theta_0) < epsilon and abs(theta_1_new - theta_1) < epsilon:
        break

    theta_0, theta_1 = theta_0_new, theta_1_new

theta_0, theta_1
````

````python
# Gr谩fica del ajuste
plt.figure()
plt.scatter(X, y, label="Datos")
plt.plot(X, theta_0 + theta_1 * X, color="red", label="Regresi贸n lineal")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
````
