## **distancia entre census tracts (hervesine)**

```python
import ast
import re
import numpy as np

def parse_geom(x):
    """
    Convierte el campo Geometry a np.array (N,2) con columnas [lon, lat].
    Soporta:
      - string tipo '[[lon,lat],[lon,lat],...]'
      - lista ya parseada
      - fallback con regex si el string viene raro
    """
    if isinstance(x, list):
        arr = np.array(x, dtype=float)
        if arr.ndim == 2 and arr.shape[1] == 2:
            return arr
        return None

    if not isinstance(x, str):
        return None

    s = x.strip()

    # intento 1: literal_eval
    try:
        obj = ast.literal_eval(s)
        arr = np.array(obj, dtype=float)
        if arr.ndim == 2 and arr.shape[1] == 2:
            return arr
    except Exception:
        pass

    # intento 2: regex por si hay basura
    pairs = re.findall(r"\[\s*([-\d\.]+)\s*,\s*([-\d\.]+)\s*\]", s)
    if not pairs:
        return None
    arr = np.array([(float(a), float(b)) for a, b in pairs], dtype=float)
    return arr

def centroid_lonlat(coords):
    """
    Centroide simple: promedio de los vertices.
    coords: np.array Nx2 [lon, lat]
    """
    if coords is None or len(coords) == 0:
        return np.nan, np.nan
    lon = float(np.mean(coords[:, 0]))
    lat = float(np.mean(coords[:, 1]))
    return lon, lat

def haversine_km(lon1, lat1, lon2, lat2):
    """
    Distancia Haversine en km (vectorizado con numpy).
    """
    R = 6371.0  # km
    lon1 = np.radians(lon1)
    lat1 = np.radians(lat1)
    lon2 = np.radians(lon2)
    lat2 = np.radians(lat2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat / 2.0) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0) ** 2
    c = 2.0 * np.arcsin(np.sqrt(a))
    return R * c

df=df_bogota.copy()

# Parse geometrías
df["orig_coords"] = df["Origin Geometry"].apply(parse_geom)
df["dest_coords"] = df["Destination Geometry"].apply(parse_geom)

# Centroides
df[["orig_lon", "orig_lat"]] = df["orig_coords"].apply(lambda a: pd.Series(centroid_lonlat(a)))
df[["dest_lon", "dest_lat"]] = df["dest_coords"].apply(lambda a: pd.Series(centroid_lonlat(a)))

# Distancia Haversine
df["dist_km"] = haversine_km(df["orig_lon"], df["orig_lat"], df["dest_lon"], df["dest_lat"])

# Revisión rápida
df[["orig_lon","orig_lat","dest_lon","dest_lat","dist_km"]]
```

```python
import ast
from shapely.geometry import Polygon

def geom_str_to_polygon(s):
    coords = ast.literal_eval(s)  # [[lon,lat],...]
    return Polygon(coords)

# Construimos un DF de destinos únicos con su polígono
dest_polys = (
    df_clean[["Destination Movement ID", "Destination Geometry"]]
    .drop_duplicates("Destination Movement ID")
    .copy()
)

dest_polys["dest_polygon"] = dest_polys["Destination Geometry"].apply(geom_str_to_polygon)
print("Destinos únicos:", len(dest_polys))
dest_polys
```

```python
import osmnx as ox
import pandas as pd
import numpy as np
from tqdm import tqdm

# Config OSMnx (reduce logs)
ox.settings.log_console = False
ox.settings.use_cache = True

sample_n = 25
sample = dest_polys.sample(sample_n, random_state=42).copy()

def compute_densities(poly):
    """
    Devuelve:
      - street_density_km: (longitud total de calles en km) / (área km^2)
      - subway_density: (# features subway/railway) / (área km^2)  [proxy simple]
    """
    area_km2 = poly.area * (111.0**2)  # Aproximación rápida (NO exacta para geodesia, pero OK para demo)
    if area_km2 <= 0:
        return np.nan, np.nan

    # 1) Street density (drive network)
    try:
        G = ox.graph_from_polygon(poly, network_type="drive", simplify=True)
        stats = ox.stats.basic_stats(G, area=area_km2)  # area en km^2
        street_density_km = stats.get("street_density_km", np.nan)
    except Exception:
        street_density_km = np.nan

    # 2) Subway/railway density (conteo de features)
    # OSMnx permite bajar geometrías por tags
    try:
        tags = {"railway": ["subway_entrance", "station", "subway"]}
        gdf = ox.geometries_from_polygon(poly, tags)
        subway_count = len(gdf)
        subway_density = subway_count / area_km2
    except Exception:
        subway_density = np.nan

    return street_density_km, subway_density

rows = []
for _, r in tqdm(sample.iterrows(), total=len(sample)):
    sd, subd = compute_densities(r["dest_polygon"])
    rows.append((r["Destination Movement ID"], sd, subd))

dens_df = pd.DataFrame(rows, columns=["Destination Movement ID", "street_density_km", "subway_density"])
dens_df.head(), dens_df.isna().mean()
```

```python
#en vez de dejar los de densidad de calle-vias NAN (492 de 496 son non-null) ponemos la mediana
df_feat["street_density_m_per_km2"] = df_feat["street_density_m_per_km2"].fillna(df_feat["street_density_m_per_km2"].median())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RANDOM_STATE = 42

# -----------------------------
# Columnas
# -----------------------------
base_cols = [
    "dist_km",
    "range_width",
    "rel_width",
    "Destination Movement ID",
    "orig_lon","orig_lat","dest_lon","dest_lat",
]

extra_cols = base_cols + ["street_density_m_per_km2"]

# Sin info de cotas (solo exógenas espaciales + ID)
cols_cero = [
    "dist_km",
    "Destination Movement ID",
    "orig_lon","orig_lat","dest_lon","dest_lat",
]

# Sin cotas + densidad
cols_cero_den = cols_cero + ["street_density_m_per_km2"]

# -----------------------------
# Función entrenamiento/evaluación
# -----------------------------
def fit_eval_rf(df_in, feature_cols, title, random_state=RANDOM_STATE, ax=None):
    y = df_in["Mean Travel Time (Seconds)"].astype(float)
    X = df_in[feature_cols].copy()

    # Imputación simple por si hay NaNs (ej: densidad)
    for c in X.columns:
        if X[c].isna().any():
            X[c] = X[c].fillna(X[c].median())

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=random_state
    )

    rf = RandomForestRegressor(
        n_estimators=600,
        random_state=random_state,
        n_jobs=-1
    )
    rf.fit(X_train, y_train)
    pred = rf.predict(X_test)

    rmse = float(np.sqrt(mean_squared_error(y_test, pred)))
    r2 = float(r2_score(y_test, pred))

    importances = pd.Series(
        rf.feature_importances_, index=X.columns
    ).sort_values(ascending=False)

    ## Pred vs Real
    #plt.figure()
    #plt.scatter(y_test, pred, alpha=0.7)
    #plt.xlabel("Real Mean Travel Time (s)")
    #plt.ylabel("Predicted Mean Travel Time (s)")
    #plt.title(title)
    #plt.show()

    # Plot en ax (si no pasas ax, crea uno solo)
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(6, 5))

    ax.scatter(y_test, pred, alpha=0.7)
    ax.set_xlabel("Real Mean Travel Time (s)")
    ax.set_ylabel("Predicted Mean Travel Time (s)")
    ax.set_title(f"{title}\nRMSE={rmse:.1f}s, R2={r2:.3f}")

    return rmse, r2, importances


# -----------------------------
# Entrenamientos
# -----------------------------
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

rmse_cero, r2_cero, imp_cero = fit_eval_rf(df_clean, cols_cero, "RF Exógeno (sin cotas, sin densidad)", ax=axs[0,0])
rmse_cero_den, r2_cero_den, imp_cero_den = fit_eval_rf(df_feat, cols_cero_den, "RF Exógeno + Street Density (sin cotas)", ax=axs[0,1])

rmse_base, r2_base, imp_base = fit_eval_rf(df_clean, base_cols, "RF Base (con cotas, sin densidad)", ax=axs[1,0])
rmse_den,  r2_den,  imp_den  = fit_eval_rf(df_feat,  extra_cols, "RF Base + Street Density (con cotas)", ax=axs[1,1])

plt.tight_layout()
plt.show()

# -----------------------------
# Tabla resumen
# -----------------------------
summary = pd.DataFrame([
    {"Model": "M0 Exógeno (sin cotas)",              "RMSE": rmse_cero,      "R2": r2_cero},
    {"Model": "M0+ Exógeno + Density (sin cotas)",   "RMSE": rmse_cero_den,  "R2": r2_cero_den},
    {"Model": "M1 Base (con cotas)",                 "RMSE": rmse_base,      "R2": r2_base},
    {"Model": "M1+ Base + Density (con cotas)",      "RMSE": rmse_den,       "R2": r2_den},
])

print(summary)

print("\n--- RESUMEN RÁPIDO ---")
print("M0   | RMSE:", rmse_cero,      "| R2:", r2_cero)
print("M0+  | RMSE:", rmse_cero_den,  "| R2:", r2_cero_den)
print("M1   | RMSE:", rmse_base,      "| R2:", r2_base)
print("M1+  | RMSE:", rmse_den,       "| R2:", r2_den)

# -----------------------------
# Importancias (top_n)
# -----------------------------
top_n = 8

def print_top(importances, name, n=top_n):
    print(f"\nTop features {name}:")
    print(importances.head(n))

print_top(imp_cero,     "M0 Exógeno (sin cotas)")
print_top(imp_cero_den, "M0+ Exógeno + Density (sin cotas)")
print_top(imp_base,     "M1 Base (con cotas)")
print_top(imp_den,      "M1+ Base + Density (con cotas)")


#display(pd.DataFrame({
#    "Model": ["RF Base", "RF + Density"],
#    "RMSE": [rmse_base, rmse_den],
#    "R2": [r2_base, r2_den],
#}))
```
