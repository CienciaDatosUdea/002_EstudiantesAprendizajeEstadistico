# Normal equation
Se puede encontrar una solucion exacta para theta sin necesidad de emplear el gradiente descente de la sesiones pasadas, para ellos se puede encontrar el valor minimo de theta y a partir de alli determinar el valor de theta que minimiza J.

Los pasos para esta minimizacion se dejan como tarea, y pueden ser calculados según lo siguiente:

Si J es la funcion de coste dada por:

\begin{equation}
J(\theta_1,\theta_2,\theta_3, ...,\theta_n )=\frac{1}{2m} \sum_{i = 1}^m (\Theta^{T} X - \hat{y}^{(i)})^2
\end{equation}


Demostrar que:

- $J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) = \frac{1}{2m} (\Theta ^ T X - y^T) (\Theta ^ T X - y^T)^T$

- $J= (\Theta ^T X) (\Theta ^T X)^T - 2(\Theta ^T X)Y  + Y^TY $


- $ \nabla _{\theta} J = \frac{1}{m} (2 X(X^T \Theta) -2XY)$


Para encontrar el valor minimo de \theta,  $\nabla _{\theta} J = 0$,

- $\Theta = (X^T X)^{-1} X^T y$



En este caso,tenemos que:

sea $X \in R^{n\times m}$ ,  $X^T \in R^{m\times n}$

sea $Y \in R^{m \times 1}$,  $Y^T \in R^{1 \times m}$

sea $\Theta \in R^{n \times 1}$, $\Theta^T \in R^{1 \times n}$




Para la demostracion anterior emplee las siguientes propiedades:

- $z^T z= \sum_i z_i^2$
- $a^T b = b^Ta$
- $\nabla _x b^T x = b$
- $\nabla _x  x^T A x = 2Ax$

donde a, b, x son matrices, $\nabla_x$ es la derivada respecto al vector x, y A es una matriz simétrica

## **Solucion:**

Si J es la funcion de coste dada por:

\begin{equation}
J(\theta_1,\theta_2,\theta_3, ...,\theta_n )=\frac{1}{2m} \sum_{i = 1}^m (\Theta^{T} X - \hat{y}^{(i)})^2
\end{equation}


Demostrar que:

- $J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) = \frac{1}{2m} (\Theta ^ T X - y^T) (\Theta ^ T X - y^T)^T$

- $J= (\Theta ^T X) (\Theta ^T X)^T - 2(\Theta ^T X)Y  + Y^TY $


- $ \nabla _{\theta} J = \frac{1}{m} (2 X(X^T \Theta) -2XY)$


Para encontrar el valor minimo de \theta,  $\nabla _{\theta} J = 0$,

- $\Theta = (X^T X)^{-1} X^T y$



En este caso,tenemos que:

sea $X \in R^{n\times m}$ ,  $X^T \in R^{m\times n}$

sea $Y \in R^{m \times 1}$,  $Y^T \in R^{1 \times m}$

sea $\Theta \in R^{n \times 1}$, $\Theta^T \in R^{1 \times n}$




Para la demostracion anterior emplee las siguientes propiedades:

- $z^T z= \sum_i z_i^2$
- $a^T b = b^Ta$
- $\nabla _x b^T x = b$
- $\nabla _x  x^T A x = 2Ax$

donde a, b, x son matrices, $\nabla_x$ es la derivada respecto al vector x, y A es una matriz simétrica

* dm $J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) = \frac{1}{2m} (\Theta ^ T X - y^T) (\Theta ^ T X - y^T)^T$

  1. por hipotesis del enunciado, tenemos que
      $$J(\theta_1,\theta_2,\theta_3, ...,\theta_n )=\frac{1}{2m} \sum_{i = 1}^m (\Theta^{T} X - \hat{y}^{(i)})^2$$
  2. ademas nos regalan-indican en el mismo, que usemos la propiedad
      $$z^T z= \sum_i z_i^2$$
  3. jugando-uniendo 1 y 2, suponga $z_i = \Theta^{T} X - \hat{y}^{(i)}$

      $$\Leftrightarrow z_i^2 = (\Theta^{T} X - \hat{y}^{(i)})^2$$

      $$\Leftrightarrow \sum z_i^2 = \sum (\Theta^{T} X - \hat{y}^{(i)})^2$$

      $$\Leftrightarrow z^T z = \sum z_i^2 = \sum (\Theta^{T} X - \hat{y}^{(i)})^2 = \(\Theta^{T} X - \hat{y}^T\)^T \(\Theta^{T} X - \hat{y}^T\) \hspace{5mm} \text{(por 2 y porque todo esta transpuesto)}$$

      $$\Leftrightarrow \frac{1}{2m} z^T z = \frac{1}{2m} \sum z_i^2 = \frac{1}{2m} \sum (\Theta^{T} X - \hat{y}^{(i)})^2 = \frac{1}{2m} \(\Theta^{T} X - \hat{y}^T\)^T \(\Theta^{T} X - \hat{y}^T\) $$

      $$\Leftrightarrow \frac{1}{2m} z^T z = \frac{1}{2m} \sum z_i^2 = \frac{1}{2m} \sum (\Theta^{T} X - \hat{y}^{(i)})^2 = J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\Theta^{T} X - \hat{y}^T\)^T \(\Theta^{T} X - \hat{y}^T\) \hspace{5mm} \text{(por 1)}$$

      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\Theta^{T} X - \hat{y}^T\)^T \(\Theta^{T} X - \hat{y}^T\) = \frac{1}{2m} \(X \Theta - \hat{y}\)^T \(X \Theta - \hat{y}\)$$
  



* dm $J= (\Theta ^T X) (\Theta ^T X)^T - 2(\Theta ^T X)Y  + Y^TY $

  1. Partiendo del ultimo resultado
      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(X \Theta - \hat{y}\)^T \(X \Theta - \hat{y}\)$$

      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\(X \Theta \)^T - \hat{y}^T\) \(X \Theta - \hat{y}\)$$

      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\(X \Theta\)^T X \Theta - \(X \Theta \)^T \hat{y} - \hat{y}^T X \Theta + \hat{y}^T\hat{y}\)$$

      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\(X \Theta \)^T X \Theta - 2\(X \Theta \)^T \hat{y} + \hat{y}^T\hat{y}\) = \frac{1}{2m} \(\(\Theta^T X\)^T \Theta^T X - 2\(\Theta^T X \)^T \hat{y} + \hat{y}^T\hat{y}\)$$

* dm $ \nabla _{\theta} J = \frac{1}{m} (2 X(X^T \Theta) -2XY)$

  1. continuando con el resultado del punto anterior
      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\(X \Theta \)^T X \Theta - 2\(X \Theta \)^T \hat{y} + \hat{y}^T\hat{y}\)$$

      $$\Leftrightarrow J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(\ \Theta^T X^T X \Theta - 2\(X \Theta\)^T \hat{y} + \hat{y}^T\hat{y}\)$$

      $$\Leftrightarrow \nabla _{\theta} J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(  \nabla _{\theta} (\ \Theta^T X^T X \Theta) - 2 \nabla _{\theta} \(X \Theta\)^T \hat{y} + \nabla _{\theta} \hat{y}^T\hat{y}\)$$

      $$\Leftrightarrow \nabla _{\theta} J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} \(  \nabla _{\theta} (\ \Theta^T X^T X \Theta) - 2 \nabla _{\theta} \Theta\^T X^T \hat{y} + 0 \)$$

  $$\Leftrightarrow \nabla _{\theta} J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{2m} (  2 X^T X \Theta - 2 X^T \hat{y} + 0 ) \hspace{5mm} \text{(pues del enunciado $\nabla_\Theta(\Theta^T A \Theta)=2A\Theta$ si A es simetrico. Como en nuestro caso $A=X^T X$ y $A = A^T$, entonces se cumple}$$

  $$\Leftrightarrow \nabla _{\theta} J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{m} (  X^T X \Theta - X^T \hat{y})$$

* encontrar el valor minimo de $\theta$ con $\nabla _{\theta} J = 0$, tal que le de $\Theta = (X^T X)^{-1} X^T y$

  1. Otra vez, retomando del resultado del punto anterior
      $$\Leftrightarrow \nabla _{\theta} J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{m} (  X^T X \Theta - X^T \hat{y})$$

      $$\Leftrightarrow \nabla _{\theta} J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) =\frac{1}{m} (  X^T X \Theta - X^T \hat{y}) = 0$$

      $$\Leftrightarrow X^T X \Theta - X^T \hat{y} = 0$$

      $$\Leftrightarrow X^T X \Theta = X^T \hat{y}$$

      $$\Leftrightarrow \Theta = (X^T X)^{-1} X^T \hat{y}$$

````python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 1) Datos sintéticos (plano)
# =========================
np.random.seed(0)

n_samples = 200
x1 = np.random.uniform(-5, 5, n_samples)
x2 = np.random.uniform(-5, 5, n_samples)

# Modelo verdadero: y = 2.1*x1 - 3.1*x2  (sin bias)
y = 2.1 * x1 - 3.1 * x2

# Matriz de diseño X: (m x n) = (200 x 3) con bias
X = np.column_stack([np.ones(n_samples), x1, x2])
y = y.reshape(-1, 1)  # (200 x 1)

print("Shapes -> X:", X.shape, " y:", y.shape)


# =========================
# 2) Ecuación normal
# =========================
# Versión estándar (requiere que X^T X sea invertible)
theta_inv = np.linalg.inv(X.T @ X) @ (X.T @ y)

# Versión robusta (pseudoinversa, por si X^T X es singular)
theta_pinv = np.linalg.pinv(X.T @ X) @ (X.T @ y)

print("\nTheta (inv):")
print(theta_inv)

print("\nTheta (pinv):")
print(theta_pinv)


# =========================
# 3) Comparación con theta real
# =========================
theta_true = np.array([[0.0], [2.1], [-3.1]])

print("\nTheta real:")
print(theta_true)

err_inv = np.linalg.norm(theta_inv - theta_true)
err_pinv = np.linalg.norm(theta_pinv - theta_true)

print(f"\nError ||theta_inv - theta_true||:  {err_inv:.6e}")
print(f"Error ||theta_pinv - theta_true||: {err_pinv:.6e}")


# =========================
# 4) (Opcional) Predicción y costo
# =========================
y_hat = X @ theta_inv
m = n_samples
cost = np.sum((y_hat - y) ** 2) / (2 * m)
print(f"\nCosto J(theta_inv): {cost:.6e}")


# =========================
# 5) (Opcional) Gráfico 3D puntos + plano ajustado
# =========================
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")

# Puntos
ax.scatter(x1, x2, y.flatten(), alpha=0.6)

# Plano ajustado (malla)
grid_x1 = np.linspace(x1.min(), x1.max(), 25)
grid_x2 = np.linspace(x2.min(), x2.max(), 25)
X1_m, X2_m = np.meshgrid(grid_x1, grid_x2)
Y_m = theta_inv[0, 0] + theta_inv[1, 0] * X1_m + theta_inv[2, 0] * X2_m

ax.plot_surface(X1_m, X2_m, Y_m, alpha=0.35)

ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("y")
ax.set_title("Puntos + Plano (Normal Equation)")
plt.show()
````

````python
import numpy as np
import pandas as pd
````

````python
#============caso univariado=============

# Cargar dataset Boston desde el link (en Colab funciona)
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)

data   = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# DataFrame con y (MEDV) y x (rm)
df = pd.DataFrame({"y": target, "rm": data[:, 5]})

# Construir X (bias + rm)
X = np.column_stack([np.ones(len(df)), df["rm"].values])
y = df["y"].values.reshape(-1, 1)

# Normal equation
theta = np.linalg.inv(X.T @ X) @ (X.T @ y)

print("Theta (bias, coef_rm):")
print(theta)

# Predicción simple para probar
y_hat = X @ theta

# RMSE
rmse = np.sqrt(np.mean((y_hat - y)**2))
print("RMSE:", rmse)
````

# Intepretación Probabilistica.

Supongamos que tenemos una caracteristica $x_i$ con m valores de entrenamiento, si asumimos que cada valor $y_i$ presenta una dispersión gaussiana $\epsilon_i$, cada $y_i$ podrá tener el siguiente valor:

$y^{i} = \Theta^T X^{(i)} + \epsilon_i$

Asumiendo ademas que el ruido gaussiando es aleatorio y esta distribuido de forma identica, con media cero y varianza $\sigma$, tenemos que la probabilidad de que la cantidad y tenga  dispersion $\epsilon_i$ es:
\begin{equation}
p(\epsilon^{(i)})=\frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( \epsilon^{(i)}\right)^2 }{2\sigma ^2}}
\end{equation}

Escribiendo, lo anterior en terminos de la probabilidad de obtener un valor de $y^{i}$ dado un $x^{i}$ parametrizado por $\theta$ obtenemos que:


\begin{equation}
p_i(y^{i}|x^{i};\theta)=\frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( y_i - \Theta^T X^{(i)} \right)^2 }{2\sigma ^2}}
\end{equation}



Si ausmimos independencia estadística de cada $\epsilon^{(i)}$, la probabilidad $L(\theta)$ asociada a toda la distribución de puntos viene dada por:

\begin{equation}
\cal{L}(\theta) = p(\vec{y}|X;\theta)=\prod_{i=1}^{n} p_i(y^{i}|x^{i};\theta)
\end{equation}



\begin{equation}
\cal{L}(\theta) =\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( y_i - \Theta^T X^{(i)} \right)^2 }{2\sigma ^2}}
\end{equation}

para tener la mejor estimación posible de los valores que se deben elegir de  $\theta$, se escogeran los parámetros que generan la mayor probabilidad de ocurrencia según las observaciones, es decir, aquellos valores para el cual $L(\theta)$ es máximo, si aplicamos el logaritmo natural antes de máximar tenemos que:

\begin{equation}
\ln \cal{L}(\theta) = \cal{l}(\theta) = \ln \left[\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( y_i - \Theta^T X^{(i)} \right)^2 }{2\sigma ^2}} \right]
\end{equation}


Después de un par de pasos se puede encontrar que:

\begin{equation}
\cal{l}(\theta) = n\ln \frac{1}{\sqrt{2\pi\sigma}} - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (y^{i}-\Theta^T X^{i})^2
\end{equation},

maximar $\cal{l(\theta)}$ equivale a encontrar donde  $\nabla_{\theta} \cal{l(\theta)} = 0$. Lo anterior muestra por que la elección de minimos cuadrados puede ser una buena eleccción para el analisis de los datos.
