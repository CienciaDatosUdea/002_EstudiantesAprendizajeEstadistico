# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y  | X_1  |
|--------|---|----------|
|$1$     |$Y^{1}$ | $X_1^{1}$|
|$2$     |$Y^{2}$ | $X_1^{2}$|
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$m$     |$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

2. Crear una función para calcular la función de coste.

````python
import numpy as np

def calcular_coste(X, y, theta0, theta1):
    m = len(y)
    predicciones = theta0 + theta1 * X
    error = predicciones - y
    J = (1 / (2 * m)) * np.sum(error ** 2)
    return J
````

### 5. Expresión teórica de la función de coste en el caso 1D y 2D

La función de coste mide el **error cuadrático medio** entre los valores predichos por el modelo y los valores reales.  
Su objetivo es encontrar los parámetros θ que minimicen dicho error.

---

#### Caso 1D

$$
J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (\theta_0 + \theta_1 x_i - y_i)^2
$$

#### Caso 2D

$$
J(\theta_0, \theta_1, \theta_2) = \frac{1}{2m} \sum_{i=1}^{m} (\theta_0 + \theta_1 x_{1i} + \theta_2 x_{2i} - y_i)^2
$$

---

```python
import numpy as np

def J_1D(theta0, theta1, X, y):
    m = len(y)
    return (1 / (2 * m)) * np.sum((theta0 + theta1 * X - y) ** 2)

def J_2D(theta0, theta1, theta2, X1, X2, y):
    m = len(y)
    return (1 / (2 * m)) * np.sum((theta0 + theta1 * X1 + theta2 * X2 - y) ** 2)


6. Construya un algoritmo en el que emplee el gradiente descente para determinar el minimo de una función, determine el mínimo con una error epsilon de  1E-4, pruebe su algoritmo para $f(x)= (x-4)^2$ y al menos 3 valores de $\alpha$


````python
import sympy as sp
import numpy as np

x = sp.Symbol('x')
f_expr = (x - 4)**2
f_deriv = sp.diff(f_expr, x)

def f(x_val):
    return float(f_expr.subs(x, x_val))

def f_derivada(x_val):
    return float(f_deriv.subs(x, x_val))

def gradiente_descendente(alpha, epsilon=1e-4, x_inicial=np.random.uniform(-10, 10)):
    x_val = x_inicial
    while True:
        grad = f_derivada(x_val)
        x_nuevo = x_val - alpha * grad
        if abs(x_nuevo - x_val) < epsilon:
            break
        x_val = x_nuevo
    return x_nuevo, f(x_nuevo)

alphas = [0.1, 0.01, 0.001]
for a in alphas:
    x_min, f_min = gradiente_descendente(alpha=a)
    print(f"α={a:.3f} → x_min={x_min:.6f}, f(x_min)={f_min:.6f}")
````

  - Determine el valor mínimo de la función con el metodo del gradiente descendente.

````python
import numpy as np
import matplotlib.pyplot as plt

def F(x, y):
    return np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2*x + 1 - np.exp(y))

def grad_F(x, y, h=1e-5):
    dF_dx = (F(x + h, y) - F(x - h, y)) / (2*h)
    dF_dy = (F(x, y + h) - F(x, y - h)) / (2*h)
    return dF_dx, dF_dy

def gradiente_descendente_2d(alpha, epsilon=1e-4, x0=0.0, y0=0.0, max_iter=10000):
    x, y = x0, y0
    for _ in range(max_iter):
        dx, dy = grad_F(x, y)
        x_new = x - alpha * dx
        y_new = y - alpha * dy
        if np.sqrt((x_new - x)**2 + (y_new - y)**2) < epsilon:
            break
        x, y = x_new, y_new
    return x, y, F(x, y)

x_vals = np.linspace(-3, 3, 200)
y_vals = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x_vals, y_vals)
Z = F(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('F(x, y)')
plt.show()

plt.figure()
CS = plt.contour(X, Y, Z, levels=50)
plt.colorbar(CS)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

for a in [0.1, 0.01, 0.001]:
    x_min, y_min, f_min = gradiente_descendente_2d(alpha=a, x0=1.0, y0=1.0)
    print(f"α={a:.3f} → x={x_min:.6f}, y={y_min:.6f}, F(x,y)={f_min:.6f}")
````

8. Empleando los siguientes datos :

```
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2*X + 0.02*np.random.random(100)
```

y las herramientas desarrolladas en los apartados anteriores,
construya un algorítmo que permita determinar una regresión lineal.

````python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)

def coste(X, y, theta0, theta1):
    m = len(y)
    return (1 / (2 * m)) * np.sum((theta0 + theta1 * X - y) ** 2)

def gradiente_descendente(X, y, alpha=0.1, epsilon=1e-6, max_iter=10000):
    theta0, theta1 = 0, 0
    for _ in range(max_iter):
        pred = theta0 + theta1 * X
        dtheta0 = (1 / len(y)) * np.sum(pred - y)
        dtheta1 = (1 / len(y)) * np.sum((pred - y) * X)
        new_theta0 = theta0 - alpha * dtheta0
        new_theta1 = theta1 - alpha * dtheta1
        if abs(new_theta0 - theta0) < epsilon and abs(new_theta1 - theta1) < epsilon:
            break
        theta0, theta1 = new_theta0, new_theta1
    return theta0, theta1

theta0_opt, theta1_opt = gradiente_descendente(X, y, alpha=0.5)
J_opt = coste(X, y, theta0_opt, theta1_opt)

print(f"θ₀ = {theta0_opt:.4f}, θ₁ = {theta1_opt:.4f}, J = {J_opt:.6f}")

plt.figure(figsize=(6,4))
plt.scatter(X, y, alpha=0.7)
plt.plot(X, theta0_opt + theta1_opt * X, color='red', linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regresión lineal con Gradiente Descendente")

texto = (
    f"$y = {theta0_opt:.4f} + {theta1_opt:.4f}x$\n"
    f"$J = {J_opt:.6f}$"
)

plt.text(0.65, 0.27, texto, fontsize=11, color='black',
         bbox=dict(facecolor='white', alpha=0.7, boxstyle='round,pad=0.5'))
plt.show()
````

````python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

X = np.linspace(0, 1, 100).reshape(-1, 1)
y = 0.2 + 0.2 * X.flatten() + 0.02 * np.random.random(100)

modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

theta0 = modelo.intercept_
theta1 = modelo.coef_[0]
r2 = r2_score(y, y_pred)

X_matriz = np.c_[np.ones(len(X)), X]
Theta = np.array([theta0, theta1])
Lambda = X_matriz @ Theta
J = np.mean((Lambda - y)**2) / 2

plt.figure(figsize=(6,4))
plt.scatter(X, y, alpha=0.7)
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regresión lineal con sklearn.LinearRegression()")

texto = (
    f"$y = {theta0:.4f} + {theta1:.4f}x$\n"
    f"$R^2 = {r2:.4f}$\n"
    f"$J = {J:.6f}$"
)

plt.text(0.65, 0.27, texto, fontsize=11, color='black',
         bbox=dict(facecolor='white', alpha=0.7, boxstyle='round,pad=0.5'))
plt.show()
````
