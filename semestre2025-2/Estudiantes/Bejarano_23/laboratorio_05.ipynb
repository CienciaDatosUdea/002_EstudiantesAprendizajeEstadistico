Se tiene: $$J(\Theta) = \frac{1}{2m}\sum_{i=1}^m(\Theta^Tx^{(i)}-y^{(i)})^2$$

Demostrar qué:
* $J(\theta_1, ..., \theta_n) = \frac{1}{2m}(\Theta^TX - y^T)(\Theta^TX-y^T)^T$

**Demostración:** Se observa que $\Theta \in \mathbb{R}^{n \times 1}$; $X \in \mathbb{R}^{n \times m}$ donde la $i$-ésima columba es $x^{(i)}$; $y \in \mathbb{R}^{m \times 1}$ es el vector columna de etiquetas, entonces $y^{T} \in \mathbb{R}^{1 \times m}$.

Luego se define el vector fila $r = \Theta^TX - y^T \in \mathbb{R}^{1\times m }$, donde sus componentes son $r_i = \Theta^Tx^{(i)} - y^{{(i)}}$

Ahora se puede ver que: $$rr^T = \sum_{i=1}^m r^2_i = \sum_{i=1}^m(\Theta^Tx^{(i)}-y^{(i)})$$ por lo tanto $$J(\Theta) = \frac{1}{2m}(\Theta^TX-y^T)(\Theta^TX-y^T)^T = \frac{1}{2m}\sum_{i=1}^m(\Theta^Tx^{(i)}-y^{(i)})$$

---

* $J=(\Theta^TX)(\Theta^TX)^T - 2(\Theta^TX)Y + Y^TY$

**Demostración:** Sea $A = \Theta^TX \in \mathbb{R}^{1\times m}$, $Y \in \mathbb{R}^{m\times 1}$ por lo tanto $B = Y^T \in \mathbb{R}^{1\times m}$, con esto se tiene $$J(\Theta) = \frac{1}{2m}(A-B)(A-B)^T$$ donde $$(A-B)(A-B)^T = AA^T - AB^T - BA^T + BB^T$$ Ahora se usa $B^T = Y$ y que los dos términos mixtos son escalares iguales (la traspuesta de un escalar es el mismo), es decir $AB^T = (BA^T)$. Por tanto: $$-AB^T-BA^T = -2AB^T = -2AY$$ además $AA^T = (\Theta^TX)(\Theta^TX)^T$ y $BB^T = Y^TY$.

Reemplazando se tiene: $$(A-B)(A-B)^T = (\Theta^TX)(\Theta^TX)^T-2(\Theta^TX)Y+Y^TY$$

Por último: $$J = \frac{1}{2m}(\Theta^TX)(\Theta^TX)^T-2(\Theta^TX)Y+Y^TY$$

---

* $\nabla_\theta J = \frac{1}{m}\left(2X\left(X^T\Theta\right)-2XY\right)$

**Demostración:** Se tiene que: $$J(\Theta) = \frac{1}{2m}\left(\Theta^TX-Y^T\right)\left(\Theta^TX-Y^T\right)^T$$ luego recordando que $\left(\Theta^TX\right)^T = X^T\Theta$ y $\left(Y^T\right)^T = Y$, entonces:

\begin{align*}
  J(\Theta) &= \frac{1}{2m}\left(\Theta^TX-Y^T\right)\left(\Theta^TX-Y^T\right)^T \\
  &= \frac{1}{2m}\left(\Theta^TX-Y^T\right)\left(X^T\Theta-Y\right) \\
  &= \frac{1}{2m}\left(\Theta^TXX^T\Theta-\Theta^TXY-Y^TX^T\Theta+Y^TY\right)
\end{align*}

Como $\Theta^TXY$ es un escalar y $\Theta^TXY = \left(Y^TX^T\Theta\right)$, los dos términos mixtos se suman a $2\Theta^TXY$. Además $XX^T$ es simétrica. Por lo tanto $$J(\Theta) = \frac{1}{2m}\left(\Theta^T\left(XX^T\right) - 2\Theta^T(XY) + Y^TY\right)$$

Luego aplicando:
* $\nabla_\Theta \left(\Theta^TA\Theta\right) = \left(A+A^T\right)\Theta$. Si $A$ es simétrica, entonces $2A\Theta$.

* $\nabla_\Theta(\Theta^T b) = b$

Aplicando esto con $A = XX^T$ y $b = XY$, se obtiene: $$\nabla_\Theta J(\Theta) = \frac{1}{2m}\left(2(XX^T)\Theta- 2(XY)\right)=\frac{1}{m}\left(\left(XX^T\right)\Theta-XY\right)$$



---

* $\Theta = (X^{T}X)^{-1}X^{T}y$

**Demostración:** Tomando la función de coste como: $$J(\Theta) = \frac{1}{2m}(X\Theta - y)^T(X\Theta-y)$$

Entonces usando el gradiente demostrado en el punto anterior, se tiene que para minimizar $J$, este se tiene que igualar a cero: $$\frac{1}{m}X^T(X\Theta - y) = 0 \ \Longrightarrow \ X^TX\Theta - X^Ty = 0$$

Despejando $\Theta$ se llega a: $$X^TX\Theta = X^Ty \ \Longrightarrow \ \Theta = \left(X^TX\right)^{-1}X^Ty$$ siempre que $X^TX$ sea invertible.
