# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y  | X_1  |
|--------|---|----------|
|$1$     |$Y^{1}$ | $X_1^{1}$|
|$2$     |$Y^{2}$ | $X_1^{2}$|
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$m$     |$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

2. Crear una función para calcular la función de coste.

````python
def cost(t0, t1, x, y):
    m = len(x)
    h = t0 + t1 * x
    return (1/(2*m)) * np.sum((h - y)**2)
````

# Función de coste (Regresión lineal) — Casos 1D y 2D

> Usaremos la definición indicada:
>
> $$
> J=\frac{1}{2m}\sum_{i=1}^{m}\Big((\theta_0-\theta_1 X^{(i)})-y^{(i)}\Big)^2
> $$

## 1) Caso 1D (una característica \(X\))

**Modelo:**
$$
h_\theta(x)=\theta_0-\theta_1 x
$$

**Coste (escalar):**
$$
J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m}\big(\theta_0-\theta_1 x^{(i)}-y^{(i)}\big)^2.
$$

**Forma matricial:**
$$
J(\theta)=\frac{1}{2m}\,\lVert X\theta-\mathbf{y}\rVert_2^2,
\qquad
X=[\,\mathbf{1},\ -\mathbf{x}\,],\ \theta=\begin{bmatrix}\theta_0\\ \theta_1\end{bmatrix}.
$$

---

## 2) Caso 2D (dos características \(X_1, X_2\))

**Modelo:**
$$
h_\theta(x)=\theta_0-\theta_1 x_1-\theta_2 x_2
$$

**Coste (escalar):**
$$
J(\theta_0,\theta_1,\theta_2)=\frac{1}{2m}\sum_{i=1}^{m}
\big(\theta_0-\theta_1 x_1^{(i)}-\theta_2 x_2^{(i)}-y^{(i)}\big)^2.
$$

**Forma matricial y solución cerrada:**
$$
J(\theta)=\frac{1}{2m}\,\lVert X\theta-\mathbf{y}\rVert_2^2,
\qquad
X=[\,\mathbf{1},\ -\mathbf{x}_1,\ -\mathbf{x}_2\,],
\qquad
\theta^\*=(X^\top X)^{-1}X^\top\mathbf{y}.
$$


6. Construya un algoritmo en el que emplee el gradiente descente para determinar el minimo de una función, determine el mínimo con una error epsilon de  1E-4, pruebe su algoritmo para $f(x)= (x-4)^2$ y al menos 3 valores de $\alpha$


````python

def f(x):
  return (x - 4.0)**2
def df(x):
  return 2.0*(x - 4.0)

def gd_1d(x0, alpha, eps=1e-4, max_iter=10_000):
    """
    Gradiente descendente 1D:
        x_{k+1} = x_k - alpha * f'(x_k)
    Criterio de parada: |x_{k+1}-x_k| < eps  OR  |f'(x_k)| < eps
    """
    x = float(x0)
    hist = [x]
    for k in range(max_iter):
        g = df(x)
        x_new = x - alpha * g
        hist.append(x_new)
        if abs(x_new - x) < eps or abs(g) < eps:
            break
        x = x_new
    return x_new, k + 1, np.array(hist)

np.random.seed(7)
x0 = np.random.uniform(-6, 6)  # punto inicial
eps = 1e-4
alphas = [0.03, 0.10, 0.30]    # tres tasas distintas

print(f"Punto inicial: x0 = {x0:.4f}\n")

resultados = []
historias = {}
for a in alphas:
    xmin, iters, h = gd_1d(x0, a, eps=eps)
    resultados.append((a, xmin, f(xmin), iters))
    historias[a] = h

# Tabla de resultados
for a, xm, fm, it in resultados:
    print(f"α={a:.2f}  |  x*≈{xm:.6f}  |  f(x*)≈{fm:.6e}  |  iteraciones={it}")
````

````python

plt.figure()
for a, h in historias.items():
    fk = [f(x) for x in h]
    plt.plot(range(len(fk)), fk, label=fr"$\alpha={a}$")
plt.yscale("log")
plt.xlabel("Iteración $k$")
plt.ylabel("$f(x_k)$")
plt.title(r"Convergencia de $f(x_k)$ para $f(x)=(x-4)^2$")
plt.legend(); plt.grid(True); plt.show()

best = min(resultados, key=lambda t: t[3])[0]
h = historias[best]
xs = np.linspace(min(h.min(), -2), max(h.max(), 6), 400)

plt.figure()
plt.plot(xs, f(xs))
plt.scatter(h, f(h), s=25)
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.title(fr"Trayectoria de GD con $\alpha={best}$  (x0={x0:.2f}, $\epsilon={eps}$)")
plt.grid(True); plt.show()
````

8. Empleando los siguientes datos :

```
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2*X + 0.02*np.random.random(100)
```

y las herramientas desarrolladas en los apartados anteriores,
construya un algorítmo que permita determinar una regresión lineal.


````python

np.random.seed(7)
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2*X + 0.02*np.random.random(100)

Xd = np.c_[np.ones_like(X), X]

def cost(theta, Xd, y):
    m = y.size
    r = Xd @ theta - y
    return (1/(2*m)) * np.dot(r, r)

def grad(theta, Xd, y):
    m = y.size
    r = Xd @ theta - y
    return (1/m) * (Xd.T @ r)

def fit_gd(Xd, y, lr=0.3, eps=1e-6, max_iter=50_000):
    theta = np.zeros(2)
    for _ in range(max_iter):
        g = grad(theta, Xd, y)
        step = lr * g
        theta_new = theta - step
        if np.linalg.norm(theta_new - theta, ord=2) < eps:
            theta = theta_new
            break
        theta = theta_new
    return theta

theta_gd = fit_gd(Xd, y, lr=0.3, eps=1e-8)
print(f"(GD)  θ0≈{theta_gd[0]:.6f},  θ1≈{theta_gd[1]:.6f},  J≈{cost(theta_gd, Xd, y):.6e}")

theta_ne, *_ = np.linalg.lstsq(Xd, y, rcond=None)
print(f"(NE)  θ0≈{theta_ne[0]:.6f},  θ1≈{theta_ne[1]:.6f},  J≈{cost(theta_ne, Xd, y):.6e}")
````

````python
x_line = np.linspace(0, 1, 200)
Xd_line = np.c_[np.ones_like(x_line), x_line]
y_gd = Xd_line @ theta_gd
y_ne = Xd_line @ theta_ne

plt.figure()
plt.scatter(X, y, s=18, label="Datos")
plt.plot(x_line, y_gd, label=rf"GD: $y={theta_gd[0]:.3f}+{theta_gd[1]:.3f}x$")
plt.plot(x_line, y_ne, linestyle="--", label=rf"NE: $y={theta_ne[0]:.3f}+{theta_ne[1]:.3f}x$")
plt.xlabel(r"$X$")
plt.ylabel(r"$y$")
plt.title(r"Regresión lineal: ajuste con GD y solución cerrada")
plt.legend()
plt.grid(True)
plt.show()
````
