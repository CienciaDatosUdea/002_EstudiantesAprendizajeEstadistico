# Normal equation
Se puede encontrar una solucion exacta para theta sin necesidad de emplear el gradiente descente de la sesiones pasadas, para ellos se puede encontrar el valor minimo de theta y a partir de alli determinar el valor de theta que minimiza J.

Los pasos para esta minimizacion se dejan como tarea, y pueden ser calculados según lo siguiente:

Si J es la funcion de coste dada por:

\begin{equation}
J(\theta_1,\theta_2,\theta_3, ...,\theta_n )=\frac{1}{2m} \sum_{i = 1}^m (\Theta^{T} X - \hat{y}^{(i)})^2
\end{equation}


Demostrar que:

- $J(\theta_1,\theta_2,\theta_3, ...,\theta_n ) = \frac{1}{2m} (\Theta ^ T X - y^T) (\Theta ^ T X - y^T)^T$

- $J= (\Theta ^T X) (\Theta ^T X)^T - 2(\Theta ^T X)Y  + Y^TY $


- $ \nabla _{\theta} J = \frac{1}{m} (2 X(X^T \Theta) -2XY)$


Para encontrar el valor minimo de \theta,  $\nabla _{\theta} J = 0$,

- $\Theta = (X^T X)^{-1} X^T y$



En este caso,tenemos que:

sea $X \in R^{n\times m}$ ,  $X^T \in R^{m\times n}$

sea $Y \in R^{m \times 1}$,  $Y^T \in R^{1 \times m}$

sea $\Theta \in R^{n \times 1}$, $\Theta^T \in R^{1 \times n}$




Para la demostracion anterior emplee las siguientes propiedades:

- $z^T z= \sum_i z_i^2$
- $a^T b = b^Ta$
- $\nabla _x b^T x = b$
- $\nabla _x  x^T A x = 2Ax$

donde a, b, x son matrices, $\nabla_x$ es la derivada respecto al vector x, y A es una matriz simétrica




## Demostración ecuación normal

Partimos de la función de coste (error cuadrático medio con factor $\tfrac{1}{2}$):

$J(\theta_1,\theta_2,\dots,\theta_n) = \dfrac{1}{2m} \sum_{i=1}^m \left(h_\Theta(x^{(i)}) - y^{(i)}\right)^2$

En notación matricial:

- $X$ = matriz de características  
- $\Theta$ = vector de parámetros  
- $Y$ = vector columna con las salidas  

La predicción del modelo es:

$\hat{Y} = X \Theta$

y el **vector de errores** es:

$e = \hat{Y} - Y = X\Theta - Y$

---

### 1. Escribir $J$ como producto matricial

Usamos la propiedad:

$z^T z = \sum_i z_i^2$

Tomando $z = e = X\Theta - Y$:

$e^T e = \sum_{i=1}^m e_i^2 = \sum_{i=1}^m (h_\Theta(x^{(i)}) - y^{(i)})^2$

Por tanto:

$J(\Theta) = \dfrac{1}{2m} e^T e = \dfrac{1}{2m} (X\Theta - Y)^T (X\Theta - Y)$

Si seguimos exactamente la notación de tu guía, donde se escribe algo como $(\Theta^T X - Y^T)(\Theta^T X - Y^T)^T$, es lo mismo pero usando vectores fila en lugar de columna. La idea clave es:

$J(\Theta) = \dfrac{1}{2m} (X\Theta - Y)^T (X\Theta - Y)$

---

### 2. Expansión de $J(\Theta)$

Partimos de:

$J(\Theta) = \dfrac{1}{2m} (X\Theta - Y)^T (X\Theta - Y)$

Expandimos el producto:

$(X\Theta - Y)^T (X\Theta - Y)
= (X\Theta)^T (X\Theta) - (X\Theta)^T Y - Y^T (X\Theta) + Y^T Y$

Sabemos que $(X\Theta)^T = \Theta^T X^T$, así que:

$(X\Theta)^T (X\Theta) = \Theta^T X^T X \Theta$

Los términos cruzados se pueden simplificar usando $a^T b = b^T a$:

- $(X\Theta)^T Y = (\Theta^T X^T) Y$ es un escalar  
- $Y^T (X\Theta)$ es el mismo escalar  
  $\Rightarrow (X\Theta)^T Y = Y^T(X\Theta)$

Por tanto:

$(X\Theta - Y)^T (X\Theta - Y)
= \Theta^T X^T X \Theta - 2\,Y^T X \Theta + Y^T Y$

Entonces:

$J(\Theta) = \dfrac{1}{2m} \Big( \Theta^T X^T X \Theta \;-\; 2\,Y^T X \Theta \;+\; Y^T Y \Big)$

Si quisieras escribirlo en el estilo de la guía (con $\Theta^T X$ primero), es la misma expresión reordenada usando $a^Tb = b^Ta$.

---

### 3. Cálculo del gradiente $\nabla_\Theta J$

Tomamos:

$J(\Theta) = \dfrac{1}{2m} \Big( \Theta^T X^T X \Theta \;-\; 2\,Y^T X \Theta \;+\; Y^T Y \Big)$

Definimos:

- $A = X^T X$ (matriz simétrica)
- $b = X^T Y$

Entonces:

$J(\Theta) = \dfrac{1}{2m} \Big( \Theta^T A \Theta - 2\,\Theta^T b + Y^T Y \Big)$

Usamos las propiedades que te dan:

- $\nabla_x (x^T A x) = 2 A x$ si $A$ es simétrica  
- $\nabla_x (b^T x) = b$  
- $\nabla_x (\text{constante}) = 0$

Aplicamos derivada con respecto a $\Theta$:

$\nabla_\Theta J(\Theta)
= \dfrac{1}{2m} \Big( 2 A \Theta - 2 b + 0 \Big)$

Sustituimos $A = X^T X$ y $b = X^T Y$:

$\nabla_\Theta J(\Theta)
= \dfrac{1}{2m} \Big( 2 X^T X \Theta - 2 X^T Y \Big)$

Simplificando el factor 2:

$\nabla_\Theta J(\Theta)
= \dfrac{1}{m} \Big( X^T X \Theta - X^T Y \Big)$

Esta es la forma compacta del gradiente de la función de coste para regresión lineal.

---

### 4. Ecuación normal: encontrar el mínimo de $J(\Theta)$

Para encontrar el valor de $\Theta$ que **minimiza** $J(\Theta)$, imponemos:

$\nabla_\Theta J(\Theta) = 0$

Es decir:

$\dfrac{1}{m} \Big( X^T X \Theta - X^T Y \Big) = 0$

Multiplicamos por $m$:

$X^T X \Theta - X^T Y = 0$

Reordenamos:

$X^T X \Theta = X^T Y$

Si la matriz $X^T X$ es invertible, podemos multiplicar por la inversa en ambos lados:

$\Theta = (X^T X)^{-1} X^T Y$

Esta expresión se conoce como la **ecuación normal** y nos da la solución exacta de $\Theta$ que minimiza la función de coste cuadrática:

$\boxed{\Theta = (X^T X)^{-1} X^T Y}$


1. Para los datos del laboratorio anterior aplicar la ecuacion normal.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
N = 100
x_1 = 2 * np.random.rand(N) - 1
x_2 = 3 * np.random.rand(N) - 1
y_label = 2.1 * x_1 - 3.1 * x_2
````

````python

X = np.column_stack([np.ones(N), x_1, x_2])
Y = y_label.reshape(-1, 1)

# ecuación normal
theta_normal = np.linalg.inv(X.T @ X) @ (X.T @ Y)

print("Theta por ecuación normal:")
print(theta_normal)
print("theta0 ~", theta_normal[0,0])
print("theta1 ~", theta_normal[1,0])
print("theta2 ~", theta_normal[2,0])
````

````python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

df = pd.DataFrame({"mean_":target, "rm":data[:,5]})
````

````python
# Matriz de diseño X (N x 2): [1, rm]
X_b = np.column_stack([np.ones(len(df)), df["rm"].values])
Y_b = df["mean_"].values.reshape(-1, 1)

# Ecuación normal
theta_boston = np.linalg.inv(X_b.T @ X_b) @ (X_b.T @ Y_b)

print("Theta para Boston (mean_ vs rm):")
print(theta_boston)
print("theta0 (intercepto) ~", theta_boston[0,0])
print("theta1 (coef rm)   ~", theta_boston[1,0])
````

````python
plt.scatter(df["rm"], df["mean_"], s=15, label="Datos reales")
x_line = np.linspace(df["rm"].min(), df["rm"].max(), 100)
y_line = theta_boston[0,0] + theta_boston[1,0] * x_line
plt.plot(x_line, y_line, label="Recta ajustada (normal equation)",color= 'r')

plt.xlabel("rm (average number of rooms)")
plt.ylabel("mean_ (valor medio de la casa)")
plt.title("Regresión lineal Boston Housing: mean_ vs rm")
plt.legend()
plt.grid(True)
plt.show()
````

# Intepretación Probabilistica.

Supongamos que tenemos una caracteristica $x_i$ con m valores de entrenamiento, si asumimos que cada valor $y_i$ presenta una dispersión gaussiana $\epsilon_i$, cada $y_i$ podrá tener el siguiente valor:

$y^{i} = \Theta^T X^{(i)} + \epsilon_i$

Asumiendo ademas que el ruido gaussiando es aleatorio y esta distribuido de forma identica, con media cero y varianza $\sigma$, tenemos que la probabilidad de que la cantidad y tenga  dispersion $\epsilon_i$ es:
\begin{equation}
p(\epsilon^{(i)})=\frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( \epsilon^{(i)}\right)^2 }{2\sigma ^2}}
\end{equation}

Escribiendo, lo anterior en terminos de la probabilidad de obtener un valor de $y^{i}$ dado un $x^{i}$ parametrizado por $\theta$ obtenemos que:


\begin{equation}
p_i(y^{i}|x^{i};\theta)=\frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( y_i - \Theta^T X^{(i)} \right)^2 }{2\sigma ^2}}
\end{equation}



Si ausmimos independencia estadística de cada $\epsilon^{(i)}$, la probabilidad $L(\theta)$ asociada a toda la distribución de puntos viene dada por:

\begin{equation}
\cal{L}(\theta) = p(\vec{y}|X;\theta)=\prod_{i=1}^{n} p_i(y^{i}|x^{i};\theta)
\end{equation}



\begin{equation}
\cal{L}(\theta) =\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( y_i - \Theta^T X^{(i)} \right)^2 }{2\sigma ^2}}
\end{equation}

para tener la mejor estimación posible de los valores que se deben elegir de  $\theta$, se escogeran los parámetros que generan la mayor probabilidad de ocurrencia según las observaciones, es decir, aquellos valores para el cual $L(\theta)$ es máximo, si aplicamos el logaritmo natural antes de máximar tenemos que:

\begin{equation}
\ln \cal{L}(\theta) = \cal{l}(\theta) = \ln \left[\prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma}} e^{-\frac{ \left( y_i - \Theta^T X^{(i)} \right)^2 }{2\sigma ^2}} \right]
\end{equation}


Después de un par de pasos se puede encontrar que:

\begin{equation}
\cal{l}(\theta) = n\ln \frac{1}{\sqrt{2\pi\sigma}} - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (y^{i}-\Theta^T X^{i})^2
\end{equation},

maximar $\cal{l(\theta)}$ equivale a encontrar donde  $\nabla_{\theta} \cal{l(\theta)} = 0$. Lo anterior muestra por que la elección de minimos cuadrados puede ser una buena eleccción para el analisis de los datos.
