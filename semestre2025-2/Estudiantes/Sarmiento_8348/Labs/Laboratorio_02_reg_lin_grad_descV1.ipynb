# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

# **Formulación y Visualización de la Función de Coste**

````python
# Librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
# 1. Cargar los datos
data = {
    'Y': [0, 1, 2, 3, 4, 5],
    'X_1': [0, 1, 2, 3, 4, 5]
}
df = pd.DataFrame(data)
df
````

````python
# 2. Función de coste (MSE)
def cost_function(X, y, theta):
    m = len(y)
    # h(x) = X * theta (asumiendo X ya incluye columna de unos si hay theta0)
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost
````

````python
# 3. Representación de la función de coste para theta0 = 0
X = df[['X_1']].values
y = df['Y'].values

theta1_vals = np.linspace(-1, 3, 101)
costs = [
    cost_function(X, y, np.array([t1])) 
    for t1 in theta1_vals
]

# Gráfico
plt.figure(figsize=(8, 4))
plt.plot(theta1_vals, costs, color='blue', linewidth=2)
plt.axvline(x=1, color='red', linestyle='--', label='Mínimo teórico (theta1=1)')
plt.xlabel(r'$\theta_1$')
plt.ylabel(r'Costo $J(\theta_1)$')
plt.title('Búsqueda del mínimo en la Función de Costo')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
# El mínimo está en theta1 = 1 (error 0 ya que Y = X_1)
````

````python
# Determinar valor minimo de coste

indice_minimo = np.argmin(costs)
theta1_optimo = theta1_vals[indice_minimo]
costo_minimo = costs[indice_minimo]

print(f"El valor mínimo de J es {costo_minimo:.4f}")
print(f"Se obtiene con theta1 = {theta1_optimo:.4f}")
````

````python
# Graficar la ecuación de regresion

plt.figure(figsize=(8, 5))

# Graficamos los datos originales del DataFrame (Puntos)
plt.scatter(df['X_1'], df['Y'], color='blue', label='Datos originales (DF)', s=100)

# Graficamos la recta de regresión con el theta1 óptimo
x_range = np.linspace(df['X_1'].min(), df['X_1'].max(), 10)
y_pred = theta1_optimo * x_range
plt.plot(x_range, y_pred, color='black', label=f'Regresión: $y = {theta1_optimo:.2f}x$')

plt.xlabel('X_1 (Característica)')
plt.ylabel('Y (Etiqueta)')
plt.title('Ajuste del modelo sobre los datos')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# 4. Función de coste 2D (theta0 y theta1)

from mpl_toolkits.mplot3d import Axes3D

theta0_vals = np.linspace(-5, 5, 100)
theta1_vals = np.linspace(-5, 5, 100)

t0, t1 = np.meshgrid(theta0_vals, theta1_vals)
J_vals = np.zeros(t0.shape)



# Añadimos columna de 1s a X para representar theta0
X_bias = np.c_[np.ones(len(X)), X]

for i in range(len(theta1_vals)):
    for j in range(len(theta0_vals)):
        t = np.array([t0[i,j], t1[i,j]])
        J_vals[i,j] = cost_function(X_bias, y, t)


fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(t0, t1, J_vals, cmap='viridis', alpha=0.8)
ax.contour(t0, t1, J_vals, 20, offset=-1, linestyles="solid")
ax.set_xlabel(r"$\theta_0$")
ax.set_ylabel(r"$\theta_1$")
ax.set_zlabel("J")
plt.show()
````

````python
plt.figure(figsize=(7, 6))
# El número 50 indica la cantidad de líneas de nivel
contour = plt.contour(t0, t1, J_vals, 50, cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8) # Añade el valor del error a las líneas
plt.xlabel(r"$\theta_0$")
plt.ylabel(r"$\theta_1$")
plt.title("Mapa de Contorno de la Función de Coste $J(\theta_0, \theta_1)$")
plt.plot(0, 1, 'rx', markersize=10, label='Mínimo Óptimo (0, 1)')
plt.legend()
plt.show()
````

**5. Expresión teórica para la función de coste**

Utilizamos la función de coste del **Error Cuadrático Medio (MSE)** para cuantificar el error de nuestro modelo de regresión lineal.

La definición matemática general para $m$ muestras de entrenamiento es:

$$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$$

Donde:
* $h_\theta(x^{(i)})$: es la predicción del modelo para la muestra $i$.
* $y^{(i)}$: es el valor real de la etiqueta para la muestra $i$.

---

**Caso 1: Regresión Lineal Simple sin Intercepto (1D)**
Este caso corresponde al análisis donde asumimos que la recta pasa por el origen ($\theta_0 = 0$). Solo optimizamos la pendiente $\theta_1$.

* **Hipótesis:** $h_\theta(x) = \theta_1 x_1$
* **Función de Coste:**

$$J(\theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (\theta_1 x_1^{(i)} - y^{(i)})^2$$

> **Interpretación:** Esta expresión corresponde a una parábola en función de $\theta_1$. El objetivo es hallar el $\theta_1$ que minimice la suma de los residuos al cuadrado.

---

**Caso 2: Regresión Lineal con Intercepto (2D)**
Este caso corresponde al modelo completo donde optimizamos tanto el punto de corte con el eje Y ($\theta_0$) como la pendiente ($\theta_1$).

* **Hipótesis:** $h_\theta(x) = \theta_0 + \theta_1 x_1$
* **Función de Coste:**

$$J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (\theta_0 + \theta_1 x_1^{(i)} - y^{(i)})^2$$

> **Interpretación:** Esta ecuación describe una superficie tridimensional conocida como **paraboloide elíptico**. Las curvas de nivel de esta función son elipses concéntricas alrededor del mínimo global.

# **Gradiente Descendente**

````python
# 6. Función definida anteriormente
def gradient_descent_simple(alpha, x_init, epsilon=1e-4):
    x = x_init
    history = [x]
    for i in range(1000):
        gradient = 2 * (x - 4) # Derivada de (x-4)^2
        new_x = x - alpha * gradient
        if abs(new_x - x) < epsilon:
            break
        x = new_x
        history.append(x)
    return x, history

# Parámetros
alphas = [0.01, 0.1, 0.9]
x_init = 0

# Configuración de la gráfica
plt.figure(figsize=(15, 5))
x_range = np.linspace(-1, 9, 100)
y_range = (x_range - 4)**2

for i, alpha in enumerate(alphas):
    min_x, history = gradient_descent_simple(alpha, x_init)
    
    # Subplot para cada alpha
    plt.subplot(1, 3, i+1)
    plt.plot(x_range, y_range, 'k--', alpha=0.3, label='f(x)')
    
    # Graficamos el camino tomado (historia)
    history = np.array(history)
    plt.plot(history, (history-4)**2, 'o-', label=f'Pasos: {len(history)}')
    plt.plot(min_x, (min_x-4)**2, 'r*', markersize=15, label='Mínimo')
    
    plt.title(f'Alpha = {alpha}\nMin encontrado: {min_x:.4f}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# 7. Definición de la función
func = lambda th: np.sin(1/2 * th[0]**2 - 1/4 * th[1]**2 + 3) * np.cos(2 * th[0] + 1 - np.exp(th[1]))

# Generación del mallado (Grid)
res = 100
_X = np.linspace(-2, 2, res)
_Y = np.linspace(-2, 2, res)
_Z = np.zeros((res, res))

for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x, y])

# Visualización 3D y Contorno
fig = plt.figure(figsize=(15, 6))

# Gráfico 3D
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(_X, _Y, _Z, cmap='viridis', alpha=0.7)
ax1.set_title("Superficie F(x, y)")

# Mapa de Contorno
ax2 = fig.add_subplot(122)
contour = ax2.contourf(_X, _Y, _Z, 50, cmap='viridis')
plt.colorbar(contour)
ax2.set_title("Mapa de Contornos (Vista de Pájaro)")
plt.show()
````

````python
# Punto de partida aleatorio
P = np.random.rand(2) * 4 - 2 # Rango [-2, 2]
learning_rate = 0.01
h = 0.001 # Incremento para la derivada numérica
history = [P.copy()]

for _ in range(10000):
    grad = np.zeros(2)
for it, val in enumerate(P):
    # Derivada central: (f(x+h) - f(x-h)) / (2*h)
    P_pos = P.copy()
    P_neg = P.copy()
    
    P_pos[it] += h
    P_neg[it] -= h
    
    grad[it] = (func(P_pos) - func(P_neg)) / (2 * h)
    # Actualización del parámetro (Descenso)
    P = P - learning_rate * grad
    history.append(P.copy())

print(f"Mínimo encontrado en: x = {P[0]:.4f}, y = {P[1]:.4f}")
print(f"Valor de la función en el mínimo: {func(P):.4f}")
````

A diferencia de los puntos anteriores, si se ejecuta este código varias veces, se obtendrá resultados diferentes porque la función tiene muchos valles. Dependiendo de dónde "caiga" el punto inicial, el gradiente te llevará al hueco más cercano (mínimo local), que no necesariamente es el más profundo de todos (mínimo global).

# **Modelo de _machine learning_**

````python
# 1. Generación de datos
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100) 

m = len(X)
X_bias = np.c_[np.ones(m), X] 

# 2. Inicialización
theta = np.zeros(2)
alpha = 0.1
iterations = 2000
cost_history = []

# 3. Algoritmo de Gradiente Descendente
for i in range(iterations):
    prediction = np.dot(X_bias, theta)     # Hipótesis
    error = prediction - y                 # Residuos
    gradient = (1/m) * np.dot(X_bias.T, error) 
    theta = theta - alpha * gradient       # Actualización
    
    cost = (1/(2*m)) * np.sum(error**2)
    cost_history.append(cost)

# 3.1 Recalcular la hipótesis final con el theta optimizado
# Esto asegura que la gráfica coincida exactamente con los parámetros finales
final_hypothesis = np.dot(X_bias, theta)

# 4. Visualización
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# A. Gráfico de Regresión
ax1.scatter(X, y, color='blue', alpha=0.5, label=r'Muestras ($y$)')
ax1.plot(X, final_hypothesis, color='red', linewidth=2, label=r'Hipótesis $h_\theta(x)$')

# Barras de residuos (Usando final_hypothesis, no prediction del loop)
for i in range(len(X)):
    label_res = 'Residuos ($e$)' if i == 0 else ""
    ax1.plot([X[i], X[i]], [y[i], final_hypothesis[i]], color='gray', 
             linestyle='--', linewidth=0.5, alpha=0.6, label=label_res)

ax1.set_title("Ajuste del Modelo y Residuos Visuales")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()
ax1.grid(True, alpha=0.3)

# B. Gráfico de Convergencia
ax2.plot(cost_history, color='green')
ax2.set_title("Curva de Convergencia")
ax2.set_xlabel("Iteración")
ax2.set_ylabel(r'Costo $J(\theta)$')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Theta final: {theta}")
````

````python
from sklearn.linear_model import LinearRegression

# 1. Preparar datos
# Scikit-learn requiere formato matriz (N, 1) para una sola feature
X_sklearn = X.reshape(-1, 1)

# 2. Entrenamiento (Solución Analítica / OLS)
model = LinearRegression(fit_intercept=True) 
model.fit(X_sklearn, y)

# 3. Extracción
theta0_sk = model.intercept_
theta1_sk = model.coef_[0]

# 4. Comparativa Numérica de Resultados
print(f"{'MÉTODO':<20} | {'INTERCEPTO (theta0)':<20} | {'PENDIENTE (theta1)':<20}")
print("-" * 70)
print(f"{'Gradiente Desc.':<20} | {theta[0]:<20.6f} | {theta[1]:<20.6f}")
print(f"{'Sklearn (OLS)':<20} | {theta0_sk:<20.6f} | {theta1_sk:<20.6f}")

# Cálculo del error relativo 
err_t0 = abs(theta[0] - theta0_sk)
err_t1 = abs(theta[1] - theta1_sk)
print("-" * 70)
print(f"Diferencia Absoluta  | {err_t0:<20.2e} | {err_t1:<20.2e}")

# 5. Prueba Visual de Convergencia
# Si el gradiente descendente fue bueno, solo deberíamos ver UNA línea (o dos muy pegadas)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.3, label='Muestras')

# Línea Sklearn (La "Verdad" matemática para estos datos)
plt.plot(X, model.predict(X_sklearn), color='green', linewidth=4, alpha=0.5, label='Sklearn (Analítica)')

# Línea Manual
y_pred_manual = theta[0] + theta[1] * X
plt.plot(X, y_pred_manual, color='red', linestyle='--', linewidth=2, label='Gradiente Descendente (Manual)')

plt.title("Comparativa: Aproximación (GD) vs Solución Exacta (OLS)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````
