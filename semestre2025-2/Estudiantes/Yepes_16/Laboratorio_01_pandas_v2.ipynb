# 7)

Ahora bien, definimos cada uno de los diagnosticos de forma númerica y agregamos la columna DiagnosisNumeric.

```python
df["DiagnosisNumeric"] = df["Diagnosis"].map({"B": 0, "M": 1})
df["DiagnosisNumeric"].head()
```

# 8.

Normalizamos el dataframe evitando alterar valores los cuales no deberíamos normalizar, como el Id, los diagnósticos númericos y los diagnosticos del dataframe.

```python
ColNum = df.drop(columns=["Diagnosis","Id","DiagnosisNumeric"]).columns
dfNormalized = df.copy()
dfNormalized[ColNum] = dfNormalized[ColNum].apply(
    lambda x: (x - x.mean()) / x.std()
)
dfNormalized.head()
```

# 9)

Ahora, a partir del dataframe anterior unimos las columnas comunes para crear un dataframe a partir de los promedios de cada columnas, obteniendo un nuevo datframe con los datos promediados.

```python
import re
grouped_features = {}

for col in df.columns:
    if col not in ["Id","Diagnosis", "DiagnosisNumeric"]:

        match = re.match(r'^[A-Za-z]+', col)
        if match:
            root = match.group(0)
            grouped_features.setdefault(root, []).append(col)


averages = pd.DataFrame()

for root, cols in grouped_features.items():
    averages[root + "Mean"] = dfNormalized[cols].mean(axis=1)

print(grouped_features)
averages.head()
```

# 12)

Ahora realizamos gráficos tipo `boxplot`, de allí identificamos diferentes outliers en los datos

```python
plot_data = pd.DataFrame({
    'RadiusMean': averages['RadiusMean'],
    'Diagnosis': dfNormalized['Diagnosis']
})

plot_data.boxplot(column='RadiusMean', by='Diagnosis', sym='k.', figsize=(18,6))

plt.title('Boxplot de RadiusMean por Diagnóstico')
plt.suptitle('')
plt.show()
```

Con el fin de quitar estos valores, podemos realizar diferentes métodos. Uno de ellos es utilizar el zscore con el fin de identificar valores extremos en los datos.

```python
import numpy as np
from scipy import stats
```

```python
plot_data['Zscore_RadiusMean'] = plot_data.groupby('Diagnosis')['RadiusMean'].transform(
    lambda x: np.abs(stats.zscore(x))
)

plot_data_filtered = plot_data[plot_data['Zscore_RadiusMean'] < 3].copy()

print("Tamaño original:", plot_data.shape[0])
print("Después de eliminar outliers por grupo:", plot_data_filtered.shape[0])

plot_data_filtered.boxplot(column="RadiusMean", by='Diagnosis', sym='k.', figsize=(18,6))
plt.title("Boxplot de RadiusMean por Diagnóstico (outliers eliminados por grupo)")
plt.suptitle("")
plt.show()
```

Al utilizar el zscore, podemos ver que los outliers fueron retirados, sin embargo uno de los valores siguen estando, esto debido a que el método utiliza media y desviación estándar, llegando a distorsionarse cuando hay valores tan extremos.

Con el fin de retirar completamente los outliers, podemos usar el siguiente método.

En este método, utilizamos los valores de los cuartiles, con el fin de eliminar los valores extremos de arriba y de abajo, así analizamos ambos grupos de datos y limpiamos los resultados.

```python

plot_data_filtered = plot_data.copy()

cleaned_groups = []

for diag, group in plot_data_filtered.groupby("Diagnosis"):
    Q1 = group["RadiusMean"].quantile(0.25)
    Q3 = group["RadiusMean"].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    group_clean = group[(group["RadiusMean"] >= lower) & (group["RadiusMean"] <= upper)]
    cleaned_groups.append(group_clean)

plot_data_filtered = pd.concat(cleaned_groups)

print("Tamaño original:", plot_data.shape[0])
print("Después de eliminar outliers por grupo (IQR):", plot_data_filtered.shape[0])

plot_data_filtered.boxplot(column="RadiusMean", by="Diagnosis", sym="k.", figsize=(18,6))
plt.title("Boxplot de RadiusMean por Diagnóstico (sin outliers bajos en B)")
plt.suptitle("")
plt.show()
```

Obteniendo así un gráfico libre de outliers.
