```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=4,
                           n_informative=2, n_redundant=0,
                           random_state=0, shuffle=False)

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X, y)

print(clf.predict([[0, 0, 0, 0]]))
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

datos = load_digits()
x = datos.data
y = datos.target

x_ent, x_pru, y_ent, y_pru = train_test_split(x, y, test_size=0.2, random_state=42)

bosque = RandomForestClassifier(n_estimators=100, random_state=42)
bosque.fit(x_ent, y_ent)

y_pred = bosque.predict(x_pru)
exactitud = accuracy_score(y_pru, y_pred)

print("Exactitud:", exactitud)

plt.gray()
plt.matshow(datos.images[95])
plt.title("Etiqueta real: " + str(y[95]))
plt.show()
```

En este laboratorio se trabaja con imágenes de números escritos a mano y se entrena un modelo que toma muchas decisiones simples para aprender a reconocerlos. En lugar de usar un solo árbol, se usan varios que colaboran entre sí, lo que hace que el resultado sea más estable y confiable. Al final, el modelo se prueba con datos nuevos para ver qué tan bien reconoce los números, mostrando cómo trabajar en conjunto mejora las decisiones.
