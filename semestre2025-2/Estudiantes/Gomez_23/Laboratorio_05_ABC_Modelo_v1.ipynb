```python
import seaborn as sns
import numpy as np

ruta_archivo = "https://raw.githubusercontent.com/hernansalinas/Curso_aprendizaje_estadistico/main/datasets/Sesion_07_housing.csv"
datos = pd.read_csv(ruta_archivo)

tipos_datos = datos.dtypes
columnas_numericas = []
for nombre_columna, tipo in tipos_datos.items():
    if tipo == 'float64' or tipo == 'int64':
        columnas_numericas.append(nombre_columna)

datos_numericos = datos[columnas_numericas]

matriz_correlacion = datos_numericos.corr(method='pearson')

nombres_filas = matriz_correlacion.index.tolist()
nombres_columnas = matriz_correlacion.columns.tolist()

valores_correlacion = matriz_correlacion.values

df_correlacion = pd.DataFrame(valores_correlacion, index=nombres_filas, columns=nombres_columnas)

print("Columnas numéricas analizadas:")
print(columnas_numericas)
print()
print("Matriz de correlación calculada:")
print(df_correlacion)
print()

figura, eje = plt.subplots(figsize=(12, 10))
mapa_calor = sns.heatmap(df_correlacion, annot=True, cmap='coolwarm', center=0, linewidths=0.6, fmt=".2f")
plt.title("Matriz de Correlación de Variables Numéricas", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

correlacion_precios = df_correlacion["median_house_value"].sort_values(ascending=False)
print("Correlaciones con el valor medio de la vivienda:")
print(correlacion_precios)
print()

plt.figure(figsize=(8, 5))
sns.barplot(x=correlacion_precios.values, y=correlacion_precios.index, palette="viridis")
plt.title("Correlación de cada variable con el valor medio de la vivienda", fontsize=14)
plt.xlabel("Coeficiente de correlación")
plt.ylabel("Variable")
plt.tight_layout()
plt.show()

umbral = 0.6
correlaciones_altas = []
for i in range(len(df_correlacion.columns)):
    for j in range(i + 1, len(df_correlacion.columns)):
        valor = df_correlacion.iloc[i, j]
        if abs(valor) > umbral:
            correlaciones_altas.append((df_correlacion.columns[i], df_correlacion.columns[j], round(valor, 2)))

print("Pares de variables con correlación fuerte (>|0.6|):")
for par in correlaciones_altas:
    print(f"{par[0]}  -  {par[1]} :  {par[2]}")

pares = [(x, y) for x, y, z in correlaciones_altas]
valores = [z for x, y, z in correlaciones_altas]

if len(correlaciones_altas) > 0:
    etiquetas = [f"{x} vs {y}" for x, y in pares]
    plt.figure(figsize=(10, 6))
    sns.barplot(x=valores, y=etiquetas, palette="rocket")
    plt.title("Correlaciones Fuertes entre Variables", fontsize=14)
    plt.xlabel("Valor de correlación")
    plt.ylabel("Pares de variables")
    plt.tight_layout()
    plt.show()
```

Para que la regresión lineal sea adecuada, idealmente se cumplen ciertas condiciones:
Relación lineal entre variables independientes y la variable objetivo Las variables derivadas (habitaciones_por_hogar, personas_por_hogar) pueden introducir cierta linealidad. Sin embargo, algunas relaciones pueden ser no lineales, por ejemplo, la influencia de la ubicación (latitude y longitude) sobre el precio.
Independencia de los errores En datos geoespaciales como este, puede existir autocorrelación, porque vecindarios cercanos suelen tener precios similares.
Homoscedasticidad (varianza constante del error)El precio de las viviendas puede tener varianza no constante: casas muy caras pueden tener errores mayores.
Normalidad de los errores No es estrictamente obligatorio para predicción, pero útil para inferencia estadística.

El score de un modelo de regresión lineal nos indica el coeficiente de determinación, es decir, la proporción de la varianza de la variable objetivo que el modelo logra explicar con las características disponibles; valores cercanos a 1 indican buen ajuste, mientras que valores bajos o negativos indican que el modelo no predice mejor que la media. El mismo conjunto de datos preprocesados (imputación, escalado, variables derivadas y codificación de categorías) puede ser usado con otros modelos supervisados, como Ridge, Lasso, Random Forest, Gradient Boosting o redes neuronales, permitiendo comparar desempeño sin rehacer el preprocesamiento. Para automatizar todo el flujo de limpieza, transformación y escalado, se pueden emplear pipelines de scikit-learn, que integran imputación, codificación, generación de nuevas variables y escalado en un único objeto que recibe los datos crudos y devuelve el dataset listo para entrenamiento, facilitando reproducibilidad, pruebas con distintos modelos y validación cruzada sin repetir código manualmente.
