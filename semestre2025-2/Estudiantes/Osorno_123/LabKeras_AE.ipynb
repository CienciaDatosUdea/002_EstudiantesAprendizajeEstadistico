```python
# 1. dataset make_moons

from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow import keras
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

X1, y1 = make_moons(n_samples=500, noise=0.2, random_state=42)

n1_samples, n1_features = X1.shape

plt.scatter(X1[:,0], X1[:,1],c=y1)
plt.show()
```

```python

def build_model1():
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(16, input_dim=2, activation='sigmoid'))
    model.add(keras.layers.Dense(16, activation='sigmoid'))
    model.add(keras.layers.Dense(1, activation = 'sigmoid'))

    opt = keras.optimizers.SGD(learning_rate=0.1)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    return model



X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)

model = build_model1()
history = model.fit(X1_train, y1_train, validation_data=(X1_test, y1_test), epochs=100, verbose=1)

_, train_acc = model.evaluate(X1_train, y1_train, verbose=0)
_, test_acc = model.evaluate(X1_test, y1_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))

plt.figure(figsize=(14,7))
# Grfica de prdidas durante el entrenamiento
plt.subplot(121)
plt.title('Loss')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
# Grfica de las precisin durante el entrenamiento
plt.subplot(122)
plt.title('Accuracy')
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='test')
plt.legend()
plt.show()
```

```python
import numpy as np
from tensorflow import keras
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Cargar dataset
digits = load_digits(n_class=6)
X2, y2 = digits.data, digits.target

# Divisin train-test
X2_train, X2_test, y2_train, y2_test = train_test_split( X2, y2, test_size=0.2, random_state=42)

def build_model():
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(32, input_dim=64, activation='relu'))
    model.add(keras.layers.Dense(32, activation='relu'))
    model.add(keras.layers.Dense(9, activation='softmax'))

    model.compile(loss='sparse_categorical_crossentropy', optimizer=keras.optimizers.Adam(0.001), metrics=['accuracy'])
    return model

keras.backend.clear_session()
model = build_model()

history = model.fit( X2_train, y2_train,validation_data=(X2_test, y2_test),epochs=50,verbose=1 )

# Evaluacin del modelo
train_acc = model.evaluate(X2_train, y2_train, verbose=0)[1]
test_acc = model.evaluate(X2_test, y2_test, verbose=0)[1]
print(f'Train: {train_acc:.3f}, Test: {test_acc:.3f}')

# Grficas:
plt.figure(figsize=(14,7))

plt.subplot(121)
plt.title('Loss')
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train','test'])

plt.subplot(122)
plt.title('Accuracy')
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['train','test'])

plt.show()
```
