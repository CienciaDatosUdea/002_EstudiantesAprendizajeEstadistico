```python
#@title 1. Leer datos

import pandas as pd


df = pd.read_csv('https://archive.ics.uci.edu/static/public/17/data.csv')
df
```

```python
#@title 2. Renombrar las columnas en formato PascalCase

df.columns = df.columns.str.replace(r'[_\s]+', ' ', regex=True).str.title().str.replace(' ', '')

df.columns
```

```python
#@title 3. Informacion dataframe
print(df.head(), df.tail(), df.describe() ,df.info())
```

```python
#@title 4. Valores nulos

val_null = df.isnull().sum()
val_nan = df.isna().sum()

print('Hay',sum(val_null)+sum(val_nan),'valores nulos. Si los hubiera, para reemplazarlos cada caso se deberia analizar, pero en principio eliminaria las filas con valores nulos')
```

```python
#@title 5. Valores únicos de diagnostico

valores = df["Diagnosis"].unique()
print("Valores únicos en diagnosis:", valores)
```

```python
#@title 6. Conteo de diagnosticos por tipo

import seaborn as sns

sns.countplot(df["Diagnosis"])
```

```python
#@title 7. Representacion numerica Diagnostico

DiagnosisNumeric = []
for el in list(df["Diagnosis"]):
    if el == "M":
        DiagnosisNumeric.append(1)
    else:
        DiagnosisNumeric.append(0)

df["DiagnosisNumeric"] = DiagnosisNumeric

df[["DiagnosisNumeric", 'Diagnosis']]
```

```python
#@title 8. Normalización datos por columnas

colsnum = df.select_dtypes(include='float64').columns
df[colsnum] = df[colsnum].apply(lambda col: (col - col.mean())/col.std(), axis=0)
```

```python
#@title 9. Agrupación de columnas del mismo tipo

import re

carac=[]
for el in df.columns:
  match = re.match(r"([a-zA-Z_]+)\d+", el)
  if match and match.group(1) not in carac:
    carac.append(match.group(1))

agruparxcarac = {c: [] for c in carac}

for el in df.columns:
  match = re.match(r"([a-zA-Z_]+)\d+", el)
  if match:
      agruparxcarac[match.group(1)].append(el)

for el in carac:
  df[f'{el}Mean']=df[agruparxcarac[el]].mean(axis=1)

df.head()
```

```python
#@title 9. Agrupación de columnas del mismo tipo
df[['Radius1','Radius2','Radius3','RadiusMean']]
```

```python
#@title 10. Histograma radio medio por diagnostico

import matplotlib.pyplot as plt

mediarxMB = df.groupby('Diagnosis')['RadiusMean'].apply(list)
plt.hist(mediarxMB['M'], label='Maligno', color='blue', alpha=0.7, bins='sturges')
plt.hist(mediarxMB['B'], label='Benigno',color='orange', alpha=0.7, bins='sturges' )
plt.legend()
plt.title(' Histograma Radio medio por tipo de diagnostico')
plt.xlabel('Radio medio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#@title 11. Graficos violin

import seaborn as sns
import matplotlib.pyplot as plt

features = ["RadiusMean", "TextureMean", "PerimeterMean", "AreaMean","SmoothnessMean", "CompactnessMean", "ConcavityMean","ConcavePointsMean", "Symmetry3", "FractalDimension3"]

df_melt = df.melt(id_vars="Diagnosis", value_vars=features, var_name="feature", value_name="value")

plt.figure(figsize=(18, 6))
sns.violinplot(data=df_melt, x="feature", y="value", hue="Diagnosis", split=True, inner="quart")
plt.show()
```

```python
#@title 12. Filtro de valores atípicos

df.boxplot(column="RadiusMean", by='Diagnosis')

Q1 = df['RadiusMean'].quantile(0.25)
Q3 = df['RadiusMean'].quantile(0.75)
IQR = Q3 - Q1
df = df[((df['RadiusMean'] > (Q1 - 1.5 * IQR)) | (df['RadiusMean'] < (Q3 + 1.5 * IQR)))]
```

```python
#@title 13. Matris de correlación

colsnum_datos = df.select_dtypes(include='float64')

corr_matrix = colsnum_datos.corr()

plt.figure(figsize=(10, 9))
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.5)
plt.title("Matriz de correlación")
plt.show()
```

```python
#@title 14. Otros graficos
#Ahora seria interesnte visualizar la correlacion entre variables epecificas que hayan
#obtenido valores obtenidos de correlacion altos en la matriz anterior.
#Se pueden visualizar usando graficos de dispercion, por ejemplo:

sns.pointplot(df, x="RadiusMean", y="AreaMean", hue= 'Diagnosis', markersize=2, markers='o', linestyles='')

```

```python
#@title 14. Otros graficos

#Seaborn tiene otras funciones para greficar las ditribuciones de los datos de forma mas suave, por ejemplo:

sns.kdeplot(data=df, x='RadiusMean', fill=True, hue='Diagnosis')
```
