# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y  | X_1  |
|--------|---|----------|
|$1$     |$Y^{1}$ | $X_1^{1}$|
|$2$     |$Y^{2}$ | $X_1^{2}$|
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$.$     |$.$     | $.$      |
|$m$     |$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

````python
# Punto 1 
import pandas as pd

# Crear los datos
data = {
    "Entrenamiento": [i for i in range(6)],
    "Y": [i for i in range(6)],
    "X_1": [i for i in range(6)]
}

# Crear el DataFrame
df = pd.DataFrame(data)
df
````

````python
import numpy as np

def funcion_costo(X, Y, theta0, theta1):
    """
    Calcula la función de costo J(theta0, theta1)
    """
    m = len(Y)
    pred = theta0 + theta1 * X
    error = pred - Y
    J = (1 / (2 * m)) * np.sum(error ** 2)
    return J
````

````python

import matplotlib.pyplot as plt

X = df["X_1"].values
Y = df["Y"].values

theta0 = 0
theta1_vals = np.linspace(-1, 3, 100)
J_vals = [funcion_costo(X, Y, theta0, t1) for t1 in theta1_vals]

plt.figure(figsize=(8, 5))
plt.plot(theta1_vals, J_vals, label="J(θ₁)")
plt.xlabel("θ₁")
plt.ylabel("Función de costo J(θ₀=0, θ₁)")
plt.title("Función de costo vs θ₁")
plt.grid(True)
plt.legend()
plt.show()
````

````python
theta1_min = theta1_vals[np.argmin(J_vals)]
J_min = min(J_vals)

print(f"Valor mínimo de θ₁: {theta1_min:.3f}")
print(f"Valor mínimo de J: {J_min:.6f}")

# Graficar los datos y la línea de regresión
plt.figure(figsize=(8,5))
plt.scatter(X, Y, color="red", label="Datos")
plt.plot(X, theta0 + theta1_min * X, label=f"h(x) = {theta0:.2f} + {theta1_min:.2f}x")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresión lineal con θ₀=0 y θ₁ óptimo")
plt.legend()
plt.grid(True)
plt.show()
````

````python
theta0_vals = np.linspace(-1, 3, 100)
theta1_vals = np.linspace(-1, 3, 100)
T0, T1 = np.meshgrid(theta0_vals, theta1_vals)

J_vals = np.zeros_like(T0)

for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        J_vals[j, i] = funcion_costo(X, Y, T0[j, i], T1[j, i])

# Gráfico de superficie
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T0, T1, J_vals, cmap='viridis', alpha=0.8)
ax.set_xlabel(r"$\theta_0$")
ax.set_ylabel(r"$\theta_1$")
ax.set_zlabel("J(θ₀, θ₁)")
ax.set_title("Superficie de la función de costo")
plt.show()
````

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

### 5. Expresión teórica de la función de costo

**Caso 1D (solo θ₁, con θ₀ = 0):**

$$
J(\theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (\theta_1 x_i - y_i)^2
$$

**Caso 2D (θ₀ y θ₁):**

$$
J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (\theta_0 + \theta_1 x_i - y_i)^2
$$

````python
# 6. Gradiente descendente para f(x) = (x - 4)^2


# Definición de la función y su derivada
def f(x):
    return (x - 4)**2

def df(x):
    return 2 * (x - 4)

def gradiente_descendente_1D(x0, alpha, epsilon=1e-4, max_iter=1000):
    """
    Gradiente descendente unidimensional
    f: función
    df: derivada
    x0: valor inicial
    alpha: tasa de aprendizaje
    epsilon: tolerancia de error
    """
    x = x0
    for i in range(max_iter):
        grad = df(x)
        nuevo_x = x - alpha * grad
        if abs(nuevo_x - x) < epsilon:
            break
        x = nuevo_x
    return x, f(x), i

# Probar con distintos valores de alpha
valores_alpha = [0.01, 0.1, 0.5]
x_inicial = np.random.uniform(-10, 10)

for alpha in valores_alpha:
    minimo, f_min, iteraciones = gradiente_descendente_1D(x_inicial, alpha)
    print(f"α = {alpha:.2f} -> x* = {minimo:.5f}, f(x*) = {f_min:.5e}, iteraciones = {iteraciones}")
````

````python
x_vals = np.linspace(-2, 10, 200)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label="f(x) = (x - 4)^2", color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Función f(x) y mínimos encontrados con distintos α")

# Graficar los mínimos encontrados
for alpha in valores_alpha:
    minimo, f_min, _ = gradiente_descendente_1D(x_inicial, alpha)
    plt.scatter(minimo, f_min, label=f"α={alpha} → x*={minimo:.2f}")

plt.legend()
plt.grid(True)
plt.show()
````

````python
# 7. Función F(x, y) y su gradiente


def F(x, y):
    return np.sin(0.5*x**2 - 0.25*y**2 + 3) * np.cos(2*x + 1 - np.exp(y))

def grad_F(x, y):
    """
    Gradiente numérico de F(x, y)
    Calculado mediante diferencias finitas pequeñas
    """
    h = 1e-5
    dFx = (F(x + h, y) - F(x - h, y)) / (2*h)
    dFy = (F(x, y + h) - F(x, y - h)) / (2*h)
    return np.array([dFx, dFy])
````

````python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z = F(X, Y)

# Gráfico 3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis", alpha=0.8)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("F(x, y)")
ax.set_title("Superficie de F(x, y)")
plt.show()

# Gráfico de contorno
plt.figure(figsize=(8,6))
plt.contourf(X, Y, Z, levels=100, cmap="plasma")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Mapa de contorno de F(x, y)")
plt.colorbar(label="F(x, y)")
plt.show()
````

````python
def gradiente_descendente_2D(x0, y0, alpha=0.01, epsilon=1e-4, max_iter=10000):
    punto = np.array([x0, y0])
    for i in range(max_iter):
        grad = grad_F(punto[0], punto[1])
        nuevo_punto = punto - alpha * grad
        if np.linalg.norm(nuevo_punto - punto) < epsilon:
            break
        punto = nuevo_punto
    return punto, F(punto[0], punto[1]), i

# Probar con un punto inicial aleatorio
x0, y0 = np.random.uniform(-2, 2, size=2)
p_min, F_min, iters = gradiente_descendente_2D(x0, y0, alpha=0.01)

print(f"Punto inicial: ({x0:.3f}, {y0:.3f})")
print(f"Mínimo encontrado: ({p_min[0]:.5f}, {p_min[1]:.5f})")
print(f"F(x*, y*) = {F_min:.5f}")
print(f"Iteraciones: {iters}")
````

````python
def gradiente_descendente_rastro(x0, y0, alpha=0.01, epsilon=1e-4, max_iter=10000):
    punto = np.array([x0, y0])
    trayectoria = [punto]
    for i in range(max_iter):
        grad = grad_F(punto[0], punto[1])
        nuevo_punto = punto - alpha * grad
        trayectoria.append(nuevo_punto)
        if np.linalg.norm(nuevo_punto - punto) < epsilon:
            break
        punto = nuevo_punto
    return np.array(trayectoria)

trayectoria = gradiente_descendente_rastro(x0, y0, alpha=0.01)

plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=50, cmap="plasma")
plt.plot(trayectoria[:,0], trayectoria[:,1], marker="o", color="red", label="Trayectoria GD")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Recorrido del gradiente descendente sobre F(x, y)")
plt.legend()
plt.grid(True)
plt.show()
````

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

Sea $\Theta^T = [\theta_0,\theta_1]$

Luego:

.\\
.\\
(\theta_0 1 + \theta_1 x_1^m)\\
\end{bmatrix}
\end{equation}


Sumando sobre todos los elementos de matriz se puede encontrar la función de coste $J = \mathrm{\Lambda.mean()^2}$.

````python
# Datos de entrenamiento
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)  # datos con un poco de ruido

# Visualizar los datos
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="red", label="Datos")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Datos de entrenamiento para regresión lineal")
plt.legend()
plt.grid(True)
plt.show()
````

````python

def gradiente_descendente_lineal(X, Y, alpha=0.1, epsilon=1e-6, max_iter=10000):
    """
    Implementación de gradiente descendente para regresión lineal
    usando la funcion_costo definida anteriormente.
    """
    theta0, theta1 = 0.0, 0.0
    m = len(Y)
    historial_coste = []

    for i in range(max_iter):
        # Predicción actual
        pred = theta0 + theta1 * X
        error = pred - Y

        # Cálculo de gradientes
        d_theta0 = (1/m) * np.sum(error)
        d_theta1 = (1/m) * np.sum(error * X)

        # Actualización de parámetros
        nuevo_theta0 = theta0 - alpha * d_theta0
        nuevo_theta1 = theta1 - alpha * d_theta1

        # Guardar costo actual
        J = funcion_costo(X, Y, nuevo_theta0, nuevo_theta1)
        historial_coste.append(J)

        # Criterio de parada
        if abs(nuevo_theta0 - theta0) < epsilon and abs(nuevo_theta1 - theta1) < epsilon:
            break

        theta0, theta1 = nuevo_theta0, nuevo_theta1

    return theta0, theta1, historial_coste
````

````python
theta0, theta1, historial = gradiente_descendente_lineal(X, y, alpha=0.5)

print(f"θ₀ = {theta0:.4f}, θ₁ = {theta1:.4f}")
print(f"Costo final J = {historial[-1]:.8f}")
print(f"Iteraciones: {len(historial)}")

# Evolución del costo
plt.figure(figsize=(8,5))
plt.plot(historial)
plt.title("Convergencia del gradiente descendente")
plt.xlabel("Iteraciones")
plt.ylabel("J(θ₀, θ₁)")
plt.grid(True)
plt.show()

# Gráfico del modelo ajustado
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="red", label="Datos")
plt.plot(X, theta0 + theta1 * X, color="blue", label=f"h(x) = {theta0:.3f} + {theta1:.3f}x")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Ajuste lineal con Gradiente Descendente")
plt.legend()
plt.grid(True)
plt.show()
````

````python

from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X.reshape(-1,1), y)

theta0_sklearn = modelo.intercept_
theta1_sklearn = modelo.coef_[0]

print(f"Regresión sklearn → θ₀ = {theta0_sklearn:.4f}, θ₁ = {theta1_sklearn:.4f}")

# Comparar visualmente ambos modelos
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="red", label="Datos")
plt.plot(X, theta0 + theta1 * X, color="blue", label="Gradiente Descendente")
plt.plot(X, modelo.predict(X.reshape(-1,1)), color="green", linestyle="--", label="sklearn")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Comparación: Gradiente Descendente vs Sklearn")
plt.legend()
plt.grid(True)
plt.show()
````
