## Análisis de datos con pandas. 

Para el siguiente [dataset](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29) realizar lo siguiente:


1. Leer los datos desde una página web.




````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Definición de columnas según la descripción del dataset
columns = [
    'id', 'diagnosis',
    'radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean',
    'concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean',
    'radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se',
    'concavity_se','concave_points_se','symmetry_se','fractal_dimension_se',
    'radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst',
    'compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst'
]

# Carga del dataset desde la fuente original (UCI)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
df = pd.read_csv(url, header=None, names=columns)

# Verificación básica del tamaño y primeras filas
print("Dimensiones del DataFrame:", df.shape)
display(df.head())
````

2. Renombrar las columnas en formato PascalCase.




````python
# Renombrar columnas a formato PascalCase
df.columns = [col.title().replace('_', '') for col in df.columns]

# Verificación del cambio
print("Nuevos nombres de columnas:\n")
print(df.columns.tolist()[:10])
````


8. Normalizar cada columna respecto a su media y desviación estándar utilizando la fórmula:  
   `(x - mean(x)) / std(x)`



````python
# Selección de columnas numéricas (excluyendo las categóricas)
numericas = df.select_dtypes(include=[np.number]).columns

# Normalización con la fórmula (x - mean) / std
df_normalizado = df.copy()
df_normalizado[numericas] = (df[numericas] - df[numericas].mean()) / df[numericas].std()

# Verificación del resultado
df_normalizado[numericas].describe().loc[['mean', 'std']]
````

9. Agrupar características similares y calcular su promedio.
   Para las siguientes características:

```python
['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean', 'SmoothnessMean', 'CompactnessMean', 'ConcavityMean', 'ConcavePointsMean', "SymmetryMean", "FractalDimensionMean"]
```

Usa expresiones regulares para identificar patrones comunes, como `radius1`, `radius2`,`radius3`, etc., y calcular los promedios. Por ejemplo:

```python
re.match(r'^[a-zA-Z_]+', "holamundo12341").group(0)
```

También puedes utilizar el método `startswith()` para buscar estos patrones en los nombres de las columnas.




````python
import re

# Selección de las columnas a agrupar
caracteristicas = [
    'RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean', 'SmoothnessMean',
    'CompactnessMean', 'ConcavityMean', 'ConcavePointsMean', 'SymmetryMean', 'FractalDimensionMean'
]

# Agrupar columnas por el prefijo (por ejemplo: 'Radius', 'Texture', etc.)
grupos = {}
for col in caracteristicas:
    prefijo = re.match(r'^[A-Za-z]+', col).group(0)
    if prefijo not in grupos:
        grupos[prefijo] = []
    grupos[prefijo].append(col)

# Calcular el promedio de cada grupo
promedios = {}
for prefijo, cols in grupos.items():
    promedios[prefijo + 'Avg'] = df[cols].mean(axis=1)

# Crear un nuevo DataFrame con los promedios
df_promedios = pd.DataFrame(promedios)

# Verificación de los resultados
print(df_promedios.head())
````


12.  Determinar y eliminar los valores atípicos (`outliers`) en la columna `RadiusMean`.

Para identificarlos, construye un gráfico tipo `boxplot`:

```python
df.boxplot(column="RadiusMean", by='Diagnosis', sym='k.', figsize=(18,6))
```

Usa el rango intercuartílico (IQR) para definir los outliers. Los valores fuera del rango [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] se consideran outliers. Ejemplo de cómo eliminar estos valores:

```python
Q1 = df['edad'].quantile(0.25)
Q3 = df['edad'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['edad'] < (Q1 - 1.5 * IQR)) | (df['edad'] > (Q3 + 1.5 * IQR)))]
```

Otra alternativa es utilizar el puntaje Z para detectar `outliers`:

```python 
df = df[(np.abs(stats.zscore(df['edad'])) < 3)]
```





````python
import matplotlib.pyplot as plt
from scipy import stats

# Boxplot para visualizar posibles outliers en RadiusMean
plt.figure(figsize=(10,5))
df.boxplot(column="RadiusMean", by='Diagnosis', sym='k.', figsize=(10,5))
plt.title('Boxplot de RadiusMean por tipo de diagnóstico')
plt.suptitle('')
plt.xlabel('Diagnóstico')
plt.ylabel('RadiusMean')
plt.show()

# Identificación de outliers con el rango intercuartílico (IQR)
Q1 = df['RadiusMean'].quantile(0.25)
Q3 = df['RadiusMean'].quantile(0.75)
IQR = Q3 - Q1

# Filtro para eliminar los valores fuera del rango permitido
df_sin_outliers = df[~((df['RadiusMean'] < (Q1 - 1.5 * IQR)) | 
                       (df['RadiusMean'] > (Q3 + 1.5 * IQR)))]

print(f"Filas originales: {df.shape[0]}")
print(f"Filas después de eliminar outliers (IQR): {df_sin_outliers.shape[0]}")

# Alternativa: eliminación usando puntaje Z
df_zscore = df[(np.abs(stats.zscore(df['RadiusMean'])) < 3)]
print(f"Filas después de eliminar outliers (Z-score): {df_zscore.shape[0]}")
````

13. Encontrar la matrix de correlación, emplear el metodo corr(), dentro de seaborn buscar el metodo heatmap() para realizar un grafico de la matrix de correlación.


````python
# Seleccionar únicamente las columnas numéricas
df_num = df.select_dtypes(include=['number'])

# Calcular la matriz de correlación
corr_matrix = df_num.corr()

# Visualizar la matriz con un heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Matriz de correlación de variables numéricas')
plt.show()
````


14. ¿Que otro tipo de gráficos pueden ser realizados para entender mejor los datos?

**Pairplot**

El **pairplot** permite visualizar las relaciones entre varias variables numéricas al mismo tiempo.  
Cada gráfico muestra cómo dos variables se relacionan entre sí, mientras que la diagonal muestra sus distribuciones individuales.



````python
sns.pairplot(df, vars=['RadiusMean', 'TextureMean', 'PerimeterMean', 'AreaMean'], 
             hue='Diagnosis', palette={'B':'skyblue','M':'orange'})
plt.show()
````

Este tipo de gráfico facilita identificar patrones, correlaciones y posibles agrupamientos según el tipo de diagnóstico (Benigno o Maligno).
