# Laboratorio 13: Redes de Hopfield
Este laboratorio explora el funcionamiento básico de las redes de Hopfield.

```python
# Importar bibliotecas necesarias
import numpy as np
```

```python
# Definir la red de Hopfield
class HopfieldNetwork:
    def __init__(self, size):
        self.size = size
        self.weights = np.zeros((size, size))

    def train(self, patterns):
        for pattern in patterns:
            self.weights += np.outer(pattern, pattern)
        np.fill_diagonal(self.weights, 0)

    def predict(self, pattern):
        return np.sign(self.weights @ pattern)
```

```python
# Crear patrones de entrenamiento
patterns = [
    np.array([1, -1, 1, -1]),
    np.array([-1, 1, -1, 1])
]

# Inicializar y entrenar la red
network = HopfieldNetwork(size=4)
network.train(patterns)
```

```python
# Probar la red con un patrón ruidoso
noisy_pattern = np.array([1, -1, -1, -1])
predicted_pattern = network.predict(noisy_pattern)

print("Patrón ruidoso:", noisy_pattern)
print("Patrón corregido:", predicted_pattern)
```
