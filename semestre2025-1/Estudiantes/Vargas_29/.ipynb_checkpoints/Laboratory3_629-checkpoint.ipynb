2. Crear una función para calcular la función de coste.

```python
def cost_function(X, Y, theta0, theta1):

    m = len(Y)
    h_x = theta0 + theta1*X
    error = h_x - Y
    J = (1/(2*m)) * np.sum(error**2)

    return J
```

5. Encontrar la expresión teórica para la función de coste en el caso 1D y 2D:

Para un caso general de $n$ variables de entrada $(x_1,⋯,x_n)$, la función de tiene una forma:

\begin{equation}
J(Θ) = \frac{1}{2 m} \sum_{i = 0}^{m} \left( \theta_0 + \sum_{j=0}^{n} θ_j x_j - y_i \right)^2
\end{equation}

Por tanto para el caso 1D, es de la forma:

\begin{equation}
J(\theta_0,\theta_1) = \frac{1}{2 m} \sum_{i = 0}^{m} \left( \theta_0 + \theta_1 x_i - y_i \right)^2
\end{equation}

y para el caso 2D:

\begin{equation}
J(\theta_0,\theta_1, \theta_2) = \frac{1}{2 m} \sum_{i = 0}^{m} \left( \theta_0 + \theta_1 x_1^{(i)} + \theta_2 x_2^{(i)} - y_i \right)^2
\end{equation}

6. Construya un algoritmo en el que emplee el gradiente descente para determinar el minimo de una función, determine el mínimo con una error epsilon de  1E-4, pruebe su algoritmo para $f(x)= (x-4)^2$ y al menos 3 valores de $\alpha$

```python
def minimum_gradient_descent(f,df_dx,x_init,alpha,epsilon,iterations):
    x = x_init

    for i in range(iterations):
      x = x - alpha*df_dx(x)
      if np.abs(df_dx(x)) < epsilon:
        break
    return x
```

```python
def f_x(x):
  return (x-4)**2

def df_dx(x):
  return 2*(x-4)

e = 1e-4
m = 1000
x_01 = minimum_gradient_descent(f_x,df_dx,0.0,0.1,e,m)
x_04 = minimum_gradient_descent(f_x,df_dx,0.0,0.4,e,m)
x_06 = minimum_gradient_descent(f_x,df_dx,0.0,0.6,e,m)
x_08 = minimum_gradient_descent(f_x,df_dx,0.0,0.8,e,m)

print(x_01)
print(x_04)
print(x_06)
print(x_08)
```

```python
def minimum_gradient_descent_2d(f,delta_f,X_init,alpha,epsilon,iterations):
    X = X_init

    for i in range(iterations):
      X = X - alpha*delta_f(X[0],X[1])
      if np.linalg.norm(delta_f(X[0],X[1])) < epsilon:
        break
    return X
```

```python
def f(x,y):
  return np.sin(0.5*x**2 - 0.25*y**2 + 3)*np.cos(2*x + 1 - np.exp(y))


def delta_f(x,y):
  arg1 = 0.5*x**2 - 0.25*y**2 + 3
  arg2 = 2*x + 1 - np.exp(y)

  df_dx = x*np.cos(arg1)*np.cos(arg2) - 2*np.sin(arg1)*np.sin(arg2)
  df_dy = -y*np.cos(arg1)*np.cos(arg2)/2 + np.exp(y)*np.sin(arg1)*np.sin(arg2)

  return np.array([df_dx,df_dy])
```

8. Empleando los siguientes datos y las herramientas desarrolladas en los apartados anteriores,
construya un algorítmo que permita determinar una regresión lineal.

```python
X = np.linspace(0, 1, 100)
Y = 0.2 + 0.2*X + 0.02*np.random.random(100)
```

```python
def quadratic_error(theta_0,theta_1):
  m = len(Y)
  h_x = theta_0 + theta_1*X
  error = h_x - Y
  J = (1/(2*m)) * np.sum(error**2)
  return J

def delta_J(theta_0,theta_1):
  dJ_dtheta0 = np.sum(theta_0 + theta_1*X - Y)/len(Y)
  dJ_dtheta1 = np.sum(X*(theta_0 + theta_1*X - Y))/len(Y)
  return np.array([dJ_dtheta0,dJ_dtheta1])

params_opt = minimum_gradient_descent_2d(quadratic_error,delta_J,[1.0,1.0],0.1,1e-4,1000)
params_opt

Y_pred = params_opt[0] + params_opt[1]*X

print("Parámetros del modelo:")
print(f"Intercepto (θ₀): {params_opt[0]:.4f}")
print(f"Coeficiente (θ₁): {params_opt[1]:.4f}")
print(f"Error cuadrático medio (MSE): {np.sum(Y-Y_pred)**2/(2*len(Y)):.8f}")
```
